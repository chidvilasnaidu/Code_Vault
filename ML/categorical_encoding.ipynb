## Dimensionality Expansion – Example

### Scenario:
Dimensionality expansion commonly occurs when applying **categorical encoding techniques** such as **One-Hot Encoding** or **Ordinal Encoding** using libraries like *scikit-learn*. These techniques convert categorical data into numerical form so that machine learning algorithms can process it.

### How it works:
For example, if a categorical feature has values like **Yes** and **No**, applying One-Hot Encoding will create separate binary columns:
- One column representing **Yes**
- Another column representing **No**

This results in an increase in the number of features.

### Effect on the dataset:
If a dataset originally contains **7 features**, applying categorical encoding may expand it to **10–18 features**, depending on the number of unique categories. This increase in dimensionality can lead to:
- Larger feature space
- Higher computational cost
- Risk of **overfitting**

### Handling dimensionality expansion:
To control unnecessary feature growth, dimensionality reduction strategies can be applied.  
One common approach is using the parameter:

```python
OneHotEncoder(drop='first')


````python
Obj2=OneHotEncoder(sparse_output=False,drop='first').set_output(transform='pandas')
Smoke=Obj2.fit_transform(df.smoker.to_frame())
Smoke
````
