```python
import requests
import json
def main():
    factin=int(input("How many facts you want to create : "))
    users=int(input("How many Random user details you want to create : "))
    print('Dumping cat facts into json file .........')
    with open("C:/Users/chidv/onedrive/Desktop/Innomatics/catdata.json",'w') as file:
        json.dump(getcatfacts(factin),file)
    print('cat facts Dumped')
    print('Dumping radam user data into json file .........')
    with open("C:/Users/chidv/onedrive/Desktop/Innomatics/userdata.json",'w') as file2:
        json.dump(getrandom_users(users),file2)
    print('user data data Dumped')
        

    
def getcatfacts(n):  # Helper Funtion1
     dict2={}
     dicat={'id':[],'facts':[]}
     for i in range(1,n):
          fact=requests.get('https://catfact.ninja/fact')
          dict2[f'fact{i}']=fact.json()['fact']
     for every in dict2:
           dicat['id'].append(every)
           dicat['facts'].append(dict2[every])
     return(dicat)


def getrandom_users(n): #hepler funtion2
    
    dict1={'name':[],'city':[],'location':[],'state':[],'email':[],'phn_num':[]}
    for i in range(1,n+1):
        randusr=requests.get('https://randomuser.me/api/')
        randusr=randusr.json()
        f=randusr['results'][0]['name']['first']
        l=randusr['results'][0]['name']['last']
        data={
            'name':f+' '+l,
            'city':randusr['results'][0]['location']['city'],
            'location':randusr['results'][0]['location'],
            'state':randusr['results'][0]['location']['state'],
            'email':randusr['results'][0]['email'],
            'phn_num':randusr['results'][0]['phone']
        }
        for every in dict1:
            dict1[every].append(data[every])
    return(dict1)
main()
```
