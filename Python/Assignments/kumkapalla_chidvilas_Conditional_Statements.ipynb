```python
#Q1. Write a program to check whether the last digit of a number (entered by user) is divisible by 3 or not. 
num=int(input("enter the number :"))
# to get last digit 
#num%10 #n//10 to remove lat digit
lstdigit=num%10

if(lstdigit%3==0):
    print("last digit is divisible by 3")
else:
    print("last digit  is not divisible by 3")
```

```python
#Q2. A company decided to give a bonus of 5% of their salary to employees if his/her years of service is more than 5 years.
#Ask the user for their salary and years of service and print the net amount of that person.

i1=int(input('Enter your salary: '))
i2=int(input('Enter your years of service: '))
if(i2>=5):
    percentage=5*i1//100
    i1+=percentage
    print(i1)
else:
    print('you were not eligible: ')
```

```python
#Q3. A school has following rules for grading system:
#a. Below 25 - F grade
#b. 25 to 45 - E
#c. 45 to 50 - D
#d. 50 to 60 - C
#e. 60 to 80 - B
#f. Above 80 - A

marks=int(input("Enter your marks: "))
if(marks>80):
    print(" A grade")
elif(marks>=60 and marks<80):
    print("B grade")
elif(marks>=50 and marks<60):
    print("C grade")
elif(marks>=45 and marks<50):
    print("D grade")
elif(marks>=25 and marks<45):
    print("E grade")
else:
    print("F grade")
          
```

```python
#Q4. Write a python program that can tell you if your sugar is normal or not. Normal fasting level sugar range is 80 to 100.
#Ask the user to enter his fasting sugar level.
#If it is below 80, then print that sugar is low.
#If it is above 100, then print that it is high 
#Otherwise print that it is normal.

num=int(input("user to enter his fasting sugar level: "))
if(num<80):
    print("sugar is low ")
elif(num>100):
    print("sugar is High ")
else:
    print("sugar is Normal")
```

```python
##Q5. Write a program which will accept a word and a number from the user. Check if the word starts with a vowel and the number is even and write appropriate conditions accordingly (Case sensitivity should not matter).

#Example :
#Innoamtics, 6 --- both the conditions are satisfied
#innomatics, 5--- at least one condition is satisfied
#Ramya, 6--- at least one condition is satisfied
#ramya , 5 --- none of the conditions are satisfied

str1=input("Enter string : ")                      # slicing the string  order 
num= int(input("Enter the number :"))
str1=str1.casefold()
first=str1[0]
lst=['a','e','i','o','u']
out=first in lst
if(out is True and num%2==0):
    print('both the conditions are satisfied')
elif(out is True or num%2==0):
     print('at least one condition is satisfied')
else:
    print('none of the conditions are satisfied')
     
```

```python
#Q6. Take input from the user. Check if a given year is a leap year.
#Example: 
  ##   Enter a year:2022
   # 2022 is not a leap year
n1=int(input("Enter a year:"))
last2=n1%100
if(last2%4==0):
    print("Its a Leap Year")
else:
    print("NON Leap year")
```
