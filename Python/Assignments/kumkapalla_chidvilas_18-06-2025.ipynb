### Problem 2

```python
# count of males and count of females
# people who are earning more than 75000 , less than 75000

# for : every_element : same result
# for : index : same result
# while : index : result
```

###  Approach 1: for : every_element :

```python

import random
random.seed(32)
categories = ["male","female"]
m_count=0
f_count=0
grt=0
less=0
data = [[random.choice(categories),random.randint(50000,100000)] for x in range(5)]
print(data)
print()
for every_element in data:
    if(every_element[0]=='male'):
        m_count+=1
    else:
        f_count+=1
for every_element in data:
    if(every_element[1]>=75000):
        grt+=1
    else:
        less+=1
        
print(f"No of males are: {m_count}")
print(f"No of males are: {f_count}")
print()
print(f"No of employees earning more than 75000: {grt}")
print(f"No of employees earning less than 75000: {less}")
```

### Approach 3 while : index : result

```python
import random
random.seed(32)
categories = ["male","female"]
m_count=0
f_count=0
grt=0
less=0
data = [[random.choice(categories),random.randint(50000,100000)] for x in range(5)]
print(data)
print()
i=0
while i<len(data): 
    if(data[i][0]=='male'):
        m_count+=1
    else:
        f_count+=1
    i+=1

j=0
while j<len(data): 
    if(data[j][1]>=75000):
        grt+=1
    else:
        less+=1
    j+=1
        
print(f"No of males are: {m_count}")
print(f"No of males are: {f_count}")
print()
print(f"No of employees earning more than 75000: {grt}")
print(f"No of employees earning less than 75000: {less}")
```
