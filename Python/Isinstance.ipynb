```python
#isinstace
#is instance is a funtion that will check the given obj is related to respective class or subclass
#returns true if it is related to class or subclass else false

#mostly it is used to check the datatype of the given var 
class lamp:
    Britnesslevel=55
class candle(lamp):
    Britnesslevel=12
```

```python
ob1=candle()
ob2=lamp()
```

```python
print(isinstance(ob1,candle))
```

```python
print(isinstance(ob1,lamp))
```

```python
print(isinstance(ob2,lamp))
```

```python
print(isinstance(ob2,candle))
```
