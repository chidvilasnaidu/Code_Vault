```python
import cv2
class img_processing:
    def __init__(self,image_path):
        self.image_path = image_path

    def display(self,img):                  # we derived a funtion such it is undergoing encapsulation ,we wrapped data that is same for all methods
        cv2.imshow("window",img)             #SO WE CALL it in each method to print it
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    
    def display_original(self):
        img=cv2.imread(self.image_path)
        self.display(img)
        
   

    def display_gray(self):
        img=cv2.imread(self.image_path)
        gray = cv2.cvtColor(img,6)
        self.display(gray)
    

    def display_binary(self):
        img=cv2.imread(self.image_path)
        _,binary = cv2.threshold(img,150,220,cv2.THRESH_BINARY)
        self.display(binary)
    
    def display_horizontal_flip(self):
        img=cv2.imread(self.image_path)
        fliph=cv2.flip(img,1)
        self.display(fliph)

    def display_vertical_flip(self):
        img=cv2.imread(self.image_path)
        flipv=cv2.flip(img,0)
        self.display(flipv)
    

    def display_gray_horinzontal_flip(self):
        img=cv2.imread(self.image_path)
        gray = cv2.cvtColor(img,cv2.COLOR_BGRA2GRAY)
        fliph=cv2.flip(gray,1)
        self.display(fliph)
    

    def display_gray_vertical_flip(self):
        img=cv2.imread(self.image_path)
        gray = cv2.cvtColor(img,cv2.COLOR_BGRA2GRAY)
        flipv=cv2.flip(gray,0)
        self.display(flipv)
        

    def display_binary_horizontal_flip(self):
        img=cv2.imread(self.image_path)
        _,binary = cv2.threshold(img,150,220,cv2.THRESH_BINARY)
        liph=cv2.flip(binary,1)
        self.display(liph)

    def display_binary_vertical_flip(self):
        img=cv2.imread(self.image_path)
        _,binary = cv2.threshold(img,150,220,cv2.THRESH_BINARY)
        flipv=cv2.flip(binary,0)
        self.display(flipv)
        
obj=img_processing(r"C:\Users\chidv\Downloads\IMG-20250501-WA0010.jpg")
obj.display_original()  
        
```

```python
import cv2
cap=cv2.VideoCapture(r"D:/phn_vds/VID-20241116-WA0019.mp4") # Load the video file
while True:
    ret,frame=cap.read() 
    cv2.imshow('graduation',frame)
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
