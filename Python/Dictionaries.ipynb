```python
##7. Given a list of numbers, store how many times each number appears using a dictionary.
## Hint: Similar to a frequency counter using for num in list:.
##Use list - 
numbers = [4, 5, 6, 7, 4, 5, 4, 6, 7, 8, 8, 4]
dist={}
for num in numbers:
    if num in dist:
         dist[num]+=1
    else:
        dist[num]=1
print(dist)
```
