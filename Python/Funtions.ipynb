### to make them dynamic we use the decorators 

```python
def decorater(funtion):
    def wrapper():
        funtion()
        print(" Hope your are Doing Well")           
        print("Have a nice day,Bye..! ")
    return wrapper
```

```python
@decorater
def sayhello1():
    print(f"Good Morning,this is funtion1 !!")  
```

```python
#normal fibonacii 
# 0,1,1,2,3,5,8...
# nterms
def fibo(num):
    a,b=0,1
    for _ in range(num):
        print(a,end=' ')
        a,b=b,b+a
fibo(10)
```

```python
# fibo singe

def fib(num):
    if(num<=0):
        return 0
    elif(num==1):
        return 1
    else:
        return fib(num-1)+fib(num-2)
def nseres(n):
    for i in range(n):
        print(fib(i),end=' ')
nseres(10)
```

```python
def fibonacci_recursive(n):
    """Return the nth Fibonacci number using recursion"""
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

fibonacci_recursive(9)
```
