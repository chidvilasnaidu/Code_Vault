```python
#we use exception handeling
try:
    numerator=int(input("Enter the numerator: "))
    print(numerator//0)  
except ZeroDivisionError:
    print("Check the denomirator is given 0")
finally:                             # it will executed regardless of error
    print("You Completed task")
```

```python
#else will executed if it doenot thrown any error 
#after sucessfull completion of task

numerator=int(input("Enter the numerator: "))
denominator=int(input("Enter the denominator: "))

try:
    print(numerator//denominator)  
except ZeroDivisionError:
    print("Check the denomirator is given 0")
else:                             # it will executed only try condition satisfies and doesnot thrown an error 
    print("you are given correct inputs")
```
