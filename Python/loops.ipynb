### Q2. Given a list of fruits, print each fruit on a new line.
#### fruits = ['apple', 'banana', 'cherry', 'date']
#### instruction - use only for loop

```python
#code here
fruits = ['apple', 'banana', 'cherry', 'date']
for i in range(len(fruits)):
    print(fruits[i])
```

### Q4. Convert the following list of words to uppercase and print each.
#### words = ['hello', 'world', 'python']

```python
#code here
words = ['hello', 'world', 'python']
for i in range(len(words)):
    print(words[i].upper())
    
```

### Q5 Using FOR loop, Print each row from the given 2D list:
#### matrix = [[1, 2],[3, 4],[5, 6]]

```python
#code here
matrix = [[1, 2],[3, 4],[5, 6]]
for i in range(len(matrix)):
    print(matrix[i])
print()
for every in matrix:
    for i in range(len(every)):
        print(every[i],end=' ')
```

### Q6. Print the first element of each row from the matrix above.
#### Hint : use indexing/slicing for accessing inside a for loop.

```python
#code here
matrix = [[1, 2],[3, 4],[5, 6]]
for i in range(len(matrix)):
    print(matrix[i][0])
```
