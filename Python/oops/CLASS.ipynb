```python
# Instance method , classmethod and static method 
class dog():
    no_of_legs=4
    no_of_tails=1
    
    # instance attribute 
    def fun1(self,name):    # can be accessed by only object
        self.name=name
        return f'{self.name} have {dog.no_of_legs} legs and {dog.no_of_tails} tail'
        
    @classmethod     # can be access by class and object
    def fun2(cls):
        return f'dog has {cls.no_of_legs}'

    @staticmethod   # can be accessed with class and object
    def fun3():
        return f' this is a static funtion '
```

```python
#instance attribute call
ob1=dog()
ob2=ob1
ob1.fun1('chimtu')
print(ob2,ob1)
```

```python
#class method acall
dog.fun2()
```

```python
dog.fun3()
```

###  how to acess the instance class properties  in classmethod and static method

1.	Return properties of instancemethod
2.	Store the return value whlie creating the objects
3.	And extra argument to class and staticmethods to pass result
4.	And pass the result from instance method as arguments to class an static methods while creating those objects



```python
#Approach 1
class dog():
    no_of_legs=4
    no_of_tails=1

    # instance attribute 
    def fun1(self,name):    
        self.name=name
        return f'{self.name} have {dog.no_of_legs} legs and {dog.no_of_tails} tail'
    @classmethod     
    def fun2(cls,res):
        return f'this is a classmethod !!!!!!! and we are accessing the instance method properties::  {res}'    # here we cannot access it directly , 
                                                   
    @staticmethod   # can be accessed with class and object
    def fun3(res):
        return f' this is a staticmethod !!!!!!! and we are accessing the instance method properties::  {res}'
```

```python
obj=dog()
result=obj.fun1('tommy')
print(obj.fun2(result))
```

```python
#accessing instance varibles by static method
obj1=dog()
result=obj1.fun1('tommy')
print(obj1.fun3(result))
```

# Access Modifiers 


```python
class cls1:
    x1 = 25         # Public member
    __x2 = 15       # Private member
    _x3 = 10        # Protected member

print('\nAccessing public members with class and object\n')
print('\tCalling by class:', cls1.x1)   # ✅ Public access by class
obj = cls1()
print('\tCalling by object:', obj.x1)   # ✅ Public access by object

print('\nAccessing private members with class and object (using name mangling)\n')

print('\tCalling by class:', cls1._cls1__x2)  # ✅ Correct (name mangling)
print('\tCalling by object:', obj._cls1__x2)  # ✅ Correct

print('\nAccessing protected members with class and object\n')
print('\tCalling by class:', cls1._x3)        # ✅ Works, but discouraged outside class
print('\tCalling by object:', obj._x3)        # ✅ Works, but discouraged
```
