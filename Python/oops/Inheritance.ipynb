### Inheritance is the process of acquiring the properties and methods of one class into another. 
* A class that inherits the behavior is called the child class, and the class that provides the behavior is called the parent class.*


```python
#1. Single Inheritance	A child class inherits from one parent class.

class parent:
    name='Shiva'
    def __init__(self,IQlevel,Strength):
        self.IQlevel=IQlevel
        self.Strength=Strength
    def printStrength(self):
        return f"{parent.name} has {self.IQlevel} IQ Level and {self.Strength} Strength"
        
class child(parent):          # inhereting the propertis all from parent
    pass

```

```python
class x:
    def task1(self):
        print('I am task 1 from parent class')

class y(x):
    def task1(self): 
        super().task1()    # Using super() to call parent class method
        print('I am task 1 from child class')

# Creating actual instances
ob1 = x()
ob1.task1()  # Calls parent class method

print('\nNow child class inherited from x:')
ob2 = y()
ob2.task1()  # Calls both parent and child methods due to super()
```
