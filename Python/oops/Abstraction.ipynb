```python
# Abstraction is Hiding the data , or showing the only essential elements by hiding the internal implementing of object
```

#### Suppose your senior developer assigns you a task, but since you're a junior and still learning, they don’t want you to break the core structure of the project. So, the senior creates an abstract class with all the required method signatures. You then implement those methods in a child class. This way, abstraction helps ensure safety, structure, and consistency across the project

```python
from abc import ABC, abstractmethod
class parent(ABC):
    def __init__(self,num):
        self.num=num
    @abstractmethod
    def add5(self):
        pass
```

```python
class child(parent):
    def add5(self):
        return 5+self.num
ob=child(5)
print(ob.add5())
```

```python
from abc import ABC,abstractmethod

class sample(ABC):

    def __init__(self,image_path):
        self.image_path = image_path
    
    @abstractmethod
    def display_original(self):pass
        
    @abstractmethod
    def display_gray(self):pass
    
    @abstractmethod
    def display_binary(self):pass
    
    @abstractmethod
    def display_horizontal_flip(self):pass
    
    @abstractmethod
    def display_vertical_flip(self):pass
    
    @abstractmethod
    def display_gray_horinzontal_flip(self):pass
    
    @abstractmethod
    def display_gray_vertical_flip(self):pass
    
    @abstractmethod
    def display_binary_horizontal_flip(self):pass
    
    @abstractmethod
    def display_binary_vertical_flip(self):pass
```

```python
import cv2
class child(sample):
    def display_original(self):
        img=cv2.imread(self.image_path)
        self.display(img)
        
    def display(self,img):
        cv2.imshow("window",img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

    def display_gray(self):
        img=cv2.imread(self.image_path)
        gray = cv2.cvtColor(img,cv2.COLOR_BGRA2GRAY)
        self.display(gray)
    
    def display_binary(self):
        img=cv2.imread(self.image_path)
        _,binary = cv2.threshold(img,150,220,cv2.THRESH_BINARY)
        self.display(binary)
    
    def display_horizontal_flip(self):
        img=cv2.imread(self.image_path)
        fliph=cv2.flip(img,1)
        self.display(fliph)

    def display_vertical_flip(self):
        img=cv2.imread(self.image_path)
        flipv=cv2.flip(img,0)
        self.display(flipv)
    

    def display_gray_horinzontal_flip(self):
        img=cv2.imread(self.image_path)
        gray = cv2.cvtColor(img,cv2.COLOR_BGRA2GRAY)
        fliph=cv2.flip(gray,1)
        self.display(fliph)
    

    def display_gray_vertical_flip(self):
        img=cv2.imread(self.image_path)
        gray = cv2.cvtColor(img,cv2.COLOR_BGRA2GRAY)
        flipv=cv2.flip(gray,0)
        self.display(flipv)
        
    

    def display_binary_horizontal_flip(self):
        img=cv2.imread(self.image_path)
        _,binary = cv2.threshold(img,150,220,cv2.THRESH_BINARY)
        liph=cv2.flip(binary,1)
        self.display(liph)
    

    def display_binary_vertical_flip(self):
        
        img=cv2.imread(self.image_path)
        _,binary = cv2.threshold(img,150,220,cv2.THRESH_BINARY)
        flipv=cv2.flip(binary,0)
        self.display(flipv)
        

        
        
```
