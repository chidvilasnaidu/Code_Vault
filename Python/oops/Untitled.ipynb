```python
#operator overloading
 #(2,3),(4,5)==(6,8)

class c1:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __add__(self,other):
        return self.x+other.x,self.y+other.y
    def __sub__(self,other):
        return self.x-other.x,self.y-other.y
    def __str__(other):
        return f"chiddu {other.x} is {other.y} "
    

ob1=c1('hello','f')
ob2=c1('english','f')
```

```python
str(ob2)
```

```python
ob1-ob2
```

```python
from abc import ABC,abstractmethod

class ca(ABC):
    @abstractmethod
    def fun1(self,x=5):
        pass
class cb(ca):
    def fun1(self,x=5):
        return x+10
```

```python
x=cb()
y=ca()
x.fun1(5)
```
