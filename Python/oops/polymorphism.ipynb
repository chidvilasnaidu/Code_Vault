```python
#method overloading
class demo:      
    def add(x,y,z=0):
        return x+y+z
obj=demo
print(obj.add(2,4))
print(obj.add(2,5,4))
```

```python
#method overloading
class demo:      
    def add(*args):
        return sum(args)
obj=demo
print(obj.add(2,4))
print(obj.add(2,5,4))
```

```python
# method overriding same name same parameters but diffrent behaviour in child class
class Animal:
    def __init__(self,a):
        self.a=a
    def add(self):
        return f'a values is  {self.a}'
class dog(Animal):
    def __init__(self,a,b):
        super().__init__(a) #using the parent a and not redining the self 
        self.b=b
    def add(self):
        return f'sum values is  {self.a+self.b}'
```

### Use super().__init__() in child class whenever parent has its own __init__ logic you want to keep.

### Then the child can extend or override without breaking parent attributes.

```python
ob1=Animal(7)
print(ob1.add())
ob2=dog(17,2)
print(ob2.add())
print(ob1.add())
```

```python
#operater overloading 

class one:
    def __init__(self,x1,y1):
        self.x1=x1
        self.y1=y1
    def __add__(self,other):  # operator overloading only work in dunder methods
        return self.x1+other.x1, self.y1+other.y1
      
ob1=one(2,4)   # our aim is to add the first element in both points and second elements in both points 
ob2=one(10,7)

ob3=('a','b')
ob4=('c','d')
```
