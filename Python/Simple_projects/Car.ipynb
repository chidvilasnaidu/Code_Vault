```python
import pygame
import time
class car:
    def ignite(self,src,duration):
        pygame.init()
        pygame.mixer.init()
        pygame.mixer.music.load(src)
        pygame.mixer.music.play()
        while pygame.mixer.music.get_busy():
            time.sleep(duration)  
            pygame.mixer.music.stop()
            
    def enginestart(self,src,duration):
        print("Starting the engine.....")
        self.ignite(src,duration)

    def Horn(self,src,duration):
        print("side....plzzzzzz")
        self.ignite(src,duration)

    
    def rain(self,src,duration):
        print("vaana vaana velluvayi...")
        self.ignite(src,duration)

            
    def clean(self,src,duration):
        print("vampires are sooon")
        self.ignite(src,duration)

            
    def skid(self,src,duration):
        print("MAYDAY !!...MAYDAY !!...MAYDAY !!...")
        self.ignite(src,duration)

    def crash(self,src,duration):
        print("kapadandayya........")
        self.ignite(src,duration)
        
    def crowd(self,src,duration):
        print("call 108")
        self.ignite(src,duration)

```
