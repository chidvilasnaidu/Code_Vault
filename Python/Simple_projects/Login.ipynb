```python
#this below problem adds and loads the data from internal memoety json files
#the code belows shows howw to create and write and load the data from json files 
```

```python
#Creating data and loding it on database files
import json
import requests as rq
database={}
for i in range(1,4):
    temp=rq.get('https://randomuser.me/api/')
    var=temp.json()
    gender=var['results'][0]['gender']
    username= var['results'][0]['login']['username']
    password=var['results'][0]['login']['password']
    database[f"user{i}"]={'username':username,'password':password,'gender':gender,'Login_status':False,'blocked':False}
with open('C:/Users/chidv/Py_practice/file_handeling/database.json','w') as file:
    json.dump(database,file)
```

```python

import json
with open('C:/Users/chidv/Py_practice/file_handeling/database.json','r') as file:
    database=json.load(file)
print(database)
```
