```python
def deposit(data,note):
    add=int(input("Enter How much amount you want to deposit : "))
    data[note]['Balance']+=add
    output=data[note]['Balance']
    return output 
    
    
```

```python
def Withdrawl(data,note):
    withdraw=int(input("Enter how much you want to withdraw the money :"))
    data[note]['Balance']-=withdraw
    output=data[note]['Balance']
    return output 
    
```
