# List Comprehention

```python
#list comprehension in pyth
#defined by using []
#	Acessing Data by LC 
#[ item for item in list]

data=[85,16,'ram',True,15.89]

[item for item in data]
```

```python
#[expression for item in iterable]
data=[85,1698,77,15.89]
[item+20 for item in data]
```

```python
#[expression for item in iterable if condition]
data=[85,1692,77,15,65,32,78]
[item for item in data if item%2==0]
```

```python
#[expression1 if condition else expression2 for item in iterable]
data=[85,1692,77,15,65,32,78]
['even' if item%2==0 else'odd' for item in data]
```

```python
#Create a list of squares from 1 to 10 using list comprehension.

[item*item for item in range(1,11)]
```

```python
#From a list of numbers, create a new list containing only even numbers using list comprehension.
nums = [1, 2, 3, 4, 5, 6]
[item for item in nums if item%2==0]
```

```python
#Convert all words in a list to uppercase using list comprehension.
words = ['hello', 'world'] 
[item.upper() for item in words]
```

```python
#From a list of strings, create a list containing only those with length > 3.
words = ['hi', 'apple', 'dog', 'banana'] 
[items for items in words if len(items)>3]
```

```python
#Generate a list of tuples (number, square) for numbers from 1 to 10 using list comprehension.
 #Output: [(1, 1), (2, 4), ..., (10, 100)]

[(i,i*i) for i in range(1,11)]
```

```python
#Create a list of all vowels in a given string using list comprehension.
lst=['a', 'e', 'i', 'o', 'u']
[i for i in "beautiful" if i.lower() in lst]
```

```python
#Flatten a nested list of 2D shape using list comprehension.
lst=[[1, 2], [3, 4], [5]]
[lst[i][j] for i in range(len(lst)) for j in range(len(lst[i]))]

```

```python
#Create a list of numbers from 1 to 100 which are divisible by both 3 and 5 using list comprehension.
[i for i in range(1,101) if i%3==0 and i%5==0]
```

```python
#Given a list words = ["data", "science", "ai", "python", "ml", "statistics"], use a list comprehension to extract words with more than 4 letters.
words = ["data", "science", "ai", "python", "ml", "statistics"]
[items for items in words if len(items)>4]
```

```python
#Given a string text = "list comprehension is powerful", use list comprehension to create a list of characters where vowels are replaced by '*'
text = "list comprehension is powerful"
lst=['a', 'e', 'i', 'o', 'u']
res=['*' if i.lower() in lst else i for i in text]
''.join(res)
```

# Dictionary comprehention

```python
#defined by using {}
{ item:expression for item in range(5)}
 {  key : value for  item in range }
```

```python
# #Create a dictionary with numbers from 1 to 5 as keys and their squares as values using dictionary comprehension.

{i:i**2 for i in range(1,5)}
```

```python
#Convert the list ['a', 'b', 'c'] to a dictionary with values as their ASCII codes.
lst=['a', 'b', 'c']
{every:ord(every) for every in lst}
```

```python
#Given a string "apple", create a dictionary with characters as keys and their counts as values.
string="apple"
{st:string.count(st) for st in string}
```

```python
#from the list ['cat', 'dog', 'rat'], create a dictionary where the key is the animal name and the value is the length of the name.
lst=['cat', 'dog', 'rat']
{every:len(every) for every in lst}
```

```python
#Filter a dictionary to include only items with values greater than 10.
import random
lat=['a','b','c','d','e','f','g']

dic={every:random.randint(1,25) for every in lat}
print('normally created dictionary',dic)
filtered={every:dic[every] for every in dic if dic[every]>10 }
print('dictionary value >10 :',filtered)
```

```python
#using the f strings in dictionaries

{f'user_{i}':random.randint(10,69) for i in range(1,6)}
```

```python
#creating the nested dictionary 

{f'user_{i}':{'age':random.randint(20,35),'salary':random.randint(45000,100000)}for i in range(1,6)}
```
