## Deep Copy

```python
lst1=['1','2','3']
lst2=lst1.copy()
print('\n lst1 before updation',lst1)
print(  ' lst2 before updation',lst2)

#updating the value in lst1
lst1[2]=8  #making changes in lst1 
#lst2[2]=165  #making changes in lst2 

print('\n lst1 After updation',lst1)
print(  ' lst2 After updation',lst2)
```

## Copy in Nested lists


```python
#Shallow copy 
# .copy is used as the shallow copy in the nested lists \

lst1=[['1','2','3'],['4','5']]
lst2=lst1.copy()
print('\n lst1 before updation',lst1)
print(  ' lst2 before updation',lst2)

#updating the value in lst1
lst1[0][0]=8  #making changes in lst1  and lst2 

print('\n lst1 After updation',lst1)  #in result both lst1,lst is updated new value  so 
print(  ' lst2 After updation',lst2)   # it is a Shallow copy 
```

```python
#Deep copy in the Nested lists 

import copy
lst1=[['1','2','3'],['4','5']]
lst2=copy.deepcopy(lst1) #-> it is syntax for the deep copy in nested lists 
print('\n lst1 before updation',lst1)
print(  ' lst2 before updation',lst2)

#updating the value in lst1
lst1[0][0]=8  #making changes only in lst1 
lst2[1][0]=6  #making changes only in lst2

print('\n lst1 After updation',lst1)  #in result only respective list is only updated 
print(  ' lst2 After updation',lst2)   # it is a deep copy 
```
