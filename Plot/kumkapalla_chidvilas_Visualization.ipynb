## 1. What are the advantages of using data visualization in data analysis?


### 1. Simplifies Complex Data
-  Visualization converts large, complex datasets into graphical formats (charts, graphs, heatmaps) that are easier to interpret. Humans can quickly detect trends, patterns, and outliers visually, which would be hard to spot in raw data tables.  
- **ex:** A company analyzing sales across 100 stores can use a **heatmap** to quickly see which regions have higher sales, instead of scanning thousands of rows of data.  

### 2. Enhances Decision-Making
-  Clear visual representation of data helps stakeholders make informed decisions faster. Decision-makers can quickly identify issues, trends, or opportunities.  
- **ex:** A marketing team can use a **line chart** showing monthly website traffic and conversions to decide where to allocate ad spend. A sudden dip in traffic is immediately visible and can trigger timely action.  

### 3. Identifies Patterns and Relationships
-  Visualization helps uncover correlations, trends, and relationships between variables that might be missed in raw data.  
- **ex:** A **scatter plot** of advertising spend vs. sales revenue can reveal a positive correlation, showing that higher ad spending leads to increased sales.

## 2. What are the appropriate plots to visualize numerical data? Explain with examples.

## 1. Histogram
- Shows the distribution of a numerical variable by dividing data into bins.
- **ex:** Visualizing employee ages to see how many fall into age ranges like 20–29, 30–39, etc.

## 2. Box Plot 
- Displays summary statistics (median, quartiles, minimum, maximum, outliers).
- **ex:** Comparing salaries across departments to identify variability or extreme values.

## 3. Scatter Plot
-  Shows the relationship between two numerical variables; useful for identifying trends or correlations.
- **ex:** Plotting advertising spend vs. sales revenue to observe correlation.

## 4. Line Plot
-  Visualizes trends over time; often used for time series data.
- **ex:** Plotting monthly website traffic for the past year to observe seasonal trends.

## 5. Area Plot
- Similar to line plot but emphasizes cumulative values or total magnitude over time.
- **ex:** Visualizing total sales contribution from different product categories over months.



## 3. What are the plots used for visualizing categorical data? Provide examples


- **Bar Chart:** Represents the frequency or count of categories. The height of each bar shows the value for that category.  
  **Ex:** Visualizing the number of employees in different departments (HR, Sales, IT).

- **Column Chart:** Similar to a bar chart but oriented vertically. Useful for comparing categories.  
  **Ex:** Showing sales by product type (Electronics, Clothing, Furniture) over a month.

- **Pie Chart:** Shows the proportion of categories in a dataset as slices of a circle. Best for small numbers of categories.  
  **Ex:** Visualizing market share of 4 major smartphone brands.

- **Stacked Bar Chart:** Shows composition within categories. Each bar is divided into sub-categories.  
  **Ex:** Visualizing sales by product type across different regions. Each bar represents a region, and the segments show product type contribution.

- **Grouped Bar Chart:** Used to compare multiple sub-categories across main categories. Each main category has a group of bars representing the sub-categories.  
  **Ex 1:** Visualizing monthly sales of different products (Electronics, Furniture, Clothing) across two regions (North and South). Each month has two bars representing North and South.  
  **Ex 2:** Comparing exam scores of students in different subjects (Math, Science, English) across two classes (Class A and Class B). Each subject has two bars for Class A and Class B.

## 8What plots are suitable for time series data visualization? Explain the key parameters and provide examples.
- **1. Line Plot**

    - `x`: Time variable (dates, months, etc.) — sets the horizontal axis

## Describe box plots, including their structure, types, and use cases in data analysis. Explain where box plots are typically used and explore their main parameters.
# Box Plots

## Type of Data
- Continuous data (can be grouped by categorical data)

## Explanation
- Box plots summarize the distribution of a continuous variable using five main statistics:
  1. **Minimum** (excluding outliers)  
  2. **First Quartile (Q1 / 25th percentile)**  
  3. **Median (Q2 / 50th percentile)**  
  4. **Third Quartile (Q3 / 75th percentile)**  
  5. **Maximum** (excluding outliers)  
- Outliers are displayed as individual points outside the whiskers.
- Helps identify spread, skewness, and outliers.

## Structure
- **Box:** Represents interquartile range (IQR = Q3 - Q1)  
- **Line inside box:** Median  
- **Whiskers:** Extend from box to minimum and maximum values within 1.5×IQR  
- **Points outside whiskers:** Outliers

## Types
1. **Vertical Box Plot** – Standard orientation  
2. **Horizontal Box Plot** – Useful for comparing long category names  
3. **Grouped Box Plot** – Compare distributions across multiple categories  
4. **Notched Box Plot** – Shows confidence interval around the median

## Use Cases
- Compare distributions across groups (e.g., salary by department)  
- Identify outliers in datasets  
- Detect skewness or variability in data  
- Visualize spread and central tendency

## Key Parameters
- `x` / `y`: Variable to plot  
- `by`: Grouping variable (for grouped plots)  
- `notch`: True to add median confidence interval  
- `vert`: True for vertical, False for horizontal  
- `whis`: Determines whisker length (default 1.5 × IQR)  
- `patch_artist`: Fill box with color (True/False)  
- `showfliers`: True/False to display outliers  
- `color` / `palette`: Customize box color  
- `widths`: Box width

````python
#You have a DataFrame product_data with columns 'Product', 'Category', and 'Quantity_Sold'. Create a bar chart to visualize the quantity sold for each product category.
data = {'Product': ['A', 'B', 'C', 'D'],
            'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing'],
            'Quantity_Sold': [50, 30, 40, 20]}

colors=['grey','red','purple','green']
df=pd.DataFrame(data)
plt.figure(figsize=(12,5)) 
plt.bar(df['Category'] + " (" + df['Product'] + ")", df['Quantity_Sold'], color=colors,width=0.3)
plt.title("Quantity sold for each product category")
plt.xlabel("Category",color='Brown',fontsize=12)
plt.ylabel("Quantity_Sold",color='Brown',fontsize=12)

plt.xticks(rotation = 0)
plt.show();
````

````python
#You have a DataFrame customer_data with columns 'CustomerID', 'Age', and 'Spend'. Create a scatter plot to visualize the relationship between customer age and spending.
data = {'CustomerID': [101, 102, 103, 104, 105],
            'Age': [25, 35, 45, 30, 40],
           'Spend': [100, 150, 120, 80, 110]}
colors=['grey','red','purple','green']
df=pd.DataFrame(data)
plt.figure(figsize=(12,5)) 
plt.scatter(x=df['Age'], y=df['Spend'],marker='*',color='red',s=50)
plt.title("relationship between customer age and spending")

plt.xlabel("Age",color='Brown',fontsize=12)
plt.ylabel("Spend",color='Brown',fontsize=12)
plt.xticks(rotation = 0)
plt.show();
````

````python
#Write a Python script using Matplotlib to create a line chart representing the monthly sales data for a company. The sales data for the first half of the year (January to June) is given below:
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Units_Sold': [500, 600, 700, 800, 900, 1000]
}
plt.figure(figsize=(12,5)) 
plt.title("Monthly sales over time")
df=pd.DataFrame(data)
plt.plot(df.Month,df.Units_Sold,color='Red')

plt.xlabel("Time",color='Brown',fontsize=12)
plt.ylabel("Units_Sold",color='Brown',fontsize=12)
plt.show();
````

````python
#Develop a Python program using Matplotlib to create a horizontal bar chart representing the number of books read by students in a class. The students and their corresponding book counts are as follows:
data = {
    'Person': ['Alice', 'Bob', 'Charlie', 'Diana', 'Emma'],
    'Books_Read': [15, 20, 12, 18, 25]
}
df = pd.DataFrame(data)
plt.figure(figsize=(12,5)) 
plt.bar(df['Person'] , df['Books_Read'], color=colors,width=0.3)
plt.title("number of books read by students in a class")
plt.xlabel("Person",color='Brown',fontsize=12)
plt.ylabel("Books_Read",color='Brown',fontsize=12)

plt.xticks(rotation = 0)
plt.show();
````

````python
#Implement a Python script using Matplotlib to create a pie chart representing the distribution of expenses in a household. The expense categories and their percentages are as follows:
data = {
    'Category': ['Rent', 'Groceries', 'Utilities', 'Entertainment', 'Savings'],
    'Percentage': [30, 25, 15, 10, 20]
}
df = pd.DataFrame(data)

plt.figure(figsize=(9,9)) 
plt.title("Distribution of expenses in a household.")

plt.pie(df.Percentage,labels=df.Category,autopct="%.2f%%",explode=[0,0.2,0,0,0],shadow=0.7,radius=1)
plt.legend()
plt.show();
````

````python
#Write a Python script using Matplotlib to create a scatter plot representing the relationship between the hours studied and exam scores for students. The data is provided below:
data = {
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Exam_Scores': [60, 70, 75, 85, 90]
}
df = pd.DataFrame(data)
plt.figure(figsize=(18,8),dpi=200)
plt.title('Relationship between the hours studied and exam scores',fontsize=22)

plt.ylabel('Exam_Scores',fontsize=20,color='Brown',)
plt.xlabel('Hours_Studied',fontsize=20,color='Brown')
plt.scatter(df.Hours_Studied,df.Exam_Scores,marker='s',s=75,color='red')
plt.show();
````

````python
#Develop a Python script using Matplotlib to create a histogram of ages in a population. The ages of individuals are given below:
data = {'Ages': [18, 20, 22, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75]}
df = pd.DataFrame(data)
plt.figure(figsize=(4,5),dpi=150)
plt.title(" Ages in a population")
plt.hist(df['Ages'], bins=7, color='grey', edgecolor='white')

plt.xlabel(' Ages')
plt.ylabel('count')
plt.show();
````

````python
#Implement a Python script using Matplotlib to create a stacked column chart representing the monthly expenses of a company for two categories: Office Rent and Employee Salaries. The monthly expenses (in thousands) for the first quarter (January to March) are given below:

months = ['Month 1', 'Month 2', 'Month 3']
office_rent = [50, 60, 55]
employee_salaries = [70, 80, 75]

# Create DataFrame
df = pd.DataFrame({
    'Month': months,
    'Office Rent': office_rent,
    'Employee Salaries': employee_salaries
})

# Plot bars
plt.figure(figsize=(8,5))

# Plot total salaries
plt.bar(df['Month'], df['Employee Salaries'], color='lightgreen', label='Employee Salaries')

# Overlay office rent portion
plt.bar(df['Month'], df['Office Rent'], color='skyblue', label='Office Rent')

plt.ylabel("Amount (in thousands)")
plt.title("Monthly Expenses: Office Rent within Employee Salaries")
plt.legend()
plt.show();
````
