# Questions on the pivot table 

```python
#Count of orders by Shipping modes
Store_df.pivot_table(index='Ship Mode',values='Order ID',aggfunc='count')
```

```python
#Average discount by Product category & Sub-category

Store_df.pivot_table(index=['Category','Sub-Category'],values='Discount',aggfunc='mean')
```

```python
#Sales by Customer segment

Store_df.pivot_table(index=['Segment'],values='Sales',aggfunc='sum')
```

```python
#Sold products by Category & Sub-category

Store_df.pivot_table(index=['Category', 'Sub-Category'],values=['Product ID'],aggfunc='count')
```

```python
#Qty sold by Customer State and Product Category

Store_df.pivot_table(index=['State','Category'],values=['Quantity'],aggfunc='sum')
```

```python
#How does the average discount vary across different shipping modes and customer segments?

Store_df.pivot_table(index=['Ship Mode','Segment'],values=['Discount'],aggfunc='mean')
```

```python
#What are the top 5 most profitable cities in each region?

sub_tab=Store_df.pivot_table(index=['Region','City'],values=['Profit'],aggfunc='sum')
top5=sub_tab.groupby(level=0, group_keys=False).apply(lambda x:x.sort_values('Profit', ascending=False).head(5))
top5
```

🔹 level=0

Your index is a MultiIndex (['Region', 'City']).

level=0 means group by the first level of the index, i.e. Region.

So we’re grouping cities within each Region.

If we had written level=1, it would group by City instead (not useful here).

🔹 group_keys=False

By default, when you use .apply(), pandas adds the group key (Region name) to the result, which creates extra index nesting.

group_keys=False prevents pandas from inserting that extra level.

It keeps the output nicely formatted as Region, City → Profit.
