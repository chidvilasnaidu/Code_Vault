# loc:
* .loc[] → Label-based indexing
* You select data by the labels (names) of rows and columns.

* It includes the end index when slicing.

* Can take boolean conditions and label names.

```python
car_df.loc[:] #all rows and colums 
```

```python
car_df.loc[0:5]  #getting the row indexes 
```

```python
car_df.index=[f'row_{i}' for i in range(len(car_df))]
car_df.head()
```

```python
car_df.loc['row_1':'row_8' ] #getting the row indexes by labels
```

```python
car_df.loc['row_1':'row_8' :2] #getting the row indexes by labels with step values
```

```python
car_df.loc['row_1':'row_8' ,['make','aspiration']] #getting the multiple rows and columns
```

# .iloc[] → Position-based indexing
* You select data by the integer positions of rows and columns.

* It excludes the end index when slicing (like normal Python list slicing).

* Cannot directly take label names.

```python
car_df=pd.read_csv(r"C:\Users\chidv\Downloads\Data_sets\exported2.csv")
car_df
```

```python
car_df.iloc[::,::] # rows[start:end:step] colrows[start:end:step]
```

```python
car_df.iloc[1:5,::] #1-5 rows all columns
```

```python
car_df.iloc[0:6,0:3] #accessing 6rows ,3 columns
```

```python
car_df.iloc[0:8:2,::] #accessing rows with step values 
```

```python
car_df.iloc[0:8:2,::4] #accssing rows and colums using step values 
```

# Important!!!!!!!!!!!!!!!! **********

```python
#if you are using loc [start:end] end is inclusive -->]
car_df.loc[0:2] # 10
```

```python
#if you are using iloc [start:end) end is ecclusive-->) 
car_df.iloc[0:2] # 10
```

```python
#normal indexing or slicing it is exlusive 
car_df[1:5]['make']
```

```python
#normal indexing accesing rows and columns
car_df[1:5][['make','wheel_base']]
```

```python
#loc indexing accesing rows and columns #accessing by label names
car_df.loc[1:5,['make','wheel_base']]
```

```python
#loc indexing accesing rows and columns #accessing elements by indexing or positions 
car_df.iloc[1:5:1,0:5:2]
```


# Condition based Indexing

```python
#extracting the data by using the specic condintions
car_df[car_df.make=='audi'] #gives the data that manufacture is audi
```

### how the condition based indexing works 


```python
car_df.make=='audi'  #generates the  masked data, boolen data for any row 
```

```python
audi_cars=car_df[car_df.make=='audi'] #if you are appling the masked data return value[true/false] into data
                            #it will acts as the filter and returns the the only true values
audi_cars
```

# multiple conditions

```python
car_df[(car_df.make=='audi') & (car_df.body_style=='sedan') ] #using the and operation
```

```python
car_df[(car_df.make=='audi') & (car_df.wheel_base>100) ] #using the and operation
```

```python
car_df[(car_df.make=='audi') | (car_df.body_style=='wagon') ] #using the ORRR operation
```
