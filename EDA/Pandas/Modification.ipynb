## Two main steps to remember while changing the datatype of a column

* **Step 1:** Always make sure all the elements in the column can be converted to the new data type.  
  - Example: If the data type is `object` and values are numeric (like `"10"`, `"25"`, `"3.5"`), it is convertible to `int` or `float`.  
  - If the value inside the object is a non-numeric string (like `"abc"`), it is **not** convertible to `int` or `float` without handling it first.

* **Step 2:** `inplace` does not work here — we need to override the column name after conversion.  
  - Example:  
    ```python
    df["col_name"] = df["col_name"].astype(int)
    ```
