## Creating the DataFrame using the Arrays 

```python
array=np.array([['Amit', 'Priya', 'Rahul', 'Sneha', 'Kiran'], 
               [25, 28, 22, 30, 27],
               ['Delhi', 'Mumbai', 'Hyderabad', 'Chennai', 'Bangalore'],
               ['Male', 'Female', 'Male', 'Female', 'Male']])  # like given one is 2-D array
                                                              # it is a list inside list
```

```python
#creating the table by using the array 
table_arr=pd.DataFrame(data=array)
table_arr # so it looks like inverted array # we will transpose the array         
```

```python
table_arr.T
```

# changing the colum_labels and Row labels

* `index` : it used to change the row name 
* `column` : it used to change the column name 

```python
table_arr=pd.DataFrame(data=array.T,columns=['Name','Age','city','gender']) #changed the column name
table_arr
```

```python
table_arr=pd.DataFrame(data=array.T,columns=['Name','Age','city','gender'],index=range(1001,1006)) #changed the row name
table_arr
```

```python
# row labels and column lables will allows or acceps both string and int data type
```

```python
table_arr=pd.DataFrame(data=array.T,columns=[1,2,3,4],index=['row 1','row 2','row 3','row 4','row 5']) #changed the row name
table_arr
```

## Data Series 
- A Data Series is a one-dimensional (1-D) labeled array capable of holding any data type
- basically data series works with  is 1-D tabular data 

```python
#creating the onedimentional data

dict2={
    'Name': ['bejing', 'chigako','vishakapatnam']
}
data_series=pd.Series(dict2['Name'])
data_series
```

```python
#creating series by list
list1=['bejing', 'chigako','vishakapatnam']
ds_2=pd.Series(list1,index=range(101,104))
ds_2
```

```python
#creating series by arrays
arr1=np.array(['bejing', 'chigako','vishakapatnam'])
ds_2=pd.Series(arr1)
ds_2
```

## Escape Sequece

* ' \ ' is considered as the escape sequence for any sentence in python

suppose U want to print the statement `My name is "chidvilas"`using "" ,by default python throws error inorder to say python "" is not ending charecter,we use ' \' such that it will reconnize it as escape sequnce and avoids error

```python
print("My name is \"chidvilas\"") #->> \ make sures to tell python consider next element as string or raw text
```

```python
print('My name is \'chidvilas\'')
```
