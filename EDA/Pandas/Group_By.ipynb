# Group By






















### `groupby()` is a method used to split the data into groups based on one or more columns, and then apply aggregation or custom functions to each group.


#### - df.groupby([<categorical column(s)>])[[<numerical column(s)>]].agg_function() 
#### - df.groupby([<categorical column(s)>])[[<categorical column(s)>]].agg_function()



```python
import pandas as pd
car_df=pd.read_excel(r"C:\Users\chidv\Downloads\Data_sets\exported.xlsx")
car_df
```

```python
car_df.groupby('make') # returns the empty objects  without any operations 
```

```python
print(car_df.groupby('make')['price'])
```

```python
car_df.groupby('make')['price'].mean() #calculating the average prices of each
```

```python
#selecting the text values at right side 
car_df.groupby('body_style')['engine_location'].max()
```

```python
#selecting the text values at right side 
car_df.groupby('body_style')['engine_location'].count()
```

```python
#giving the multiple indexes(Left Side )
car_df.groupby(['make','fuel_type'])[['price']].mean()
```

```python
car_df.groupby(['make','body_style'])[["city_mpg", "highway_mpg", "price"]].mean() #multiple colums at both index and value sides
```

```python
# idf there is multiple column present at the indexes it will condider as the one tuple index so data type will be pandas series 
cards=car_df.groupby(['make','body_style'])["city_mpg"].mean() #multiple colums at both index and value sides
print(cards,'\n\n Type of cards is :',type(cards))
```

```python
#but if you give the values in the double brackets it will leads to pandas dataframe
cards=car_df.groupby(['make','body_style'])[["city_mpg"]].mean() #multiple colums at both index and value sides [[]] gives dataframe
print(cards,'\n\n Type of cards is :',type(cards))
```

```python
#combining or appling the two diff aggrigate operations on singile cvolumn 
# min & max price by make
car_df.groupby('make')['price'].agg(min_price='min',max_price='max')
```

## Assignment 


```python
# Avg city_mpg by make
# Avg city_mp by num_of_doors 
# Avg price, highway_mpg by num_of_doors, engine_location 
# count of cars by engine_location, fuel_type 
# min & max price by make
```

```python
# Avg city_mpg by make
car_df.groupby('make')[['city_mpg']].mean() #if you give the double brackets it will assumes as a pandas dataframe 
```

```python
# Avg city_mp by num_of_doors 
car_df.groupby('num_of_doors')[['city_mpg']].mean()
```

```python
# Avg price, highway_mpg by num_of_doors, engine_location 
car_df.groupby(['num_of_doors','engine_location'])[['price','highway_mpg']].mean()
```

```python
# count of cars by engine_location, fuel_type 
car_df.groupby('make')[['engine_location','fuel_type']].count()
```

```python
# min & max price by make
car_df.groupby('make')['price'].agg(min_price='min',max_price='max').reset_index()
```

# as_index

* like when you aregrouped more than one column as index left side it will acts as a index inorder to restrict those behaviour we will use as indexm


```python
carid=car_df.groupby(['make',	'fuel_type'])[['highway_mpg']].mean() #here ma
carid
```

```python
# above make	fuel_type	 acts as the index
carid.index #so multi index is there inorder to restrict
#             these act we use asindex
```

```python
car_df.groupby(['make',	'fuel_type'] ,as_index=False)[['highway_mpg']].mean()
```

```python
car_df.groupby(['make',	'fuel_type'] ,as_index=False)[['highway_mpg']].mean().index #so fixed the problem
```

## Performing multiple aggrigate operations at same time 

```python
car_subdf=car_df.groupby(['make','body_style'] ,as_index=False)[['compression_ratio','highway_mpg','price']].mean()
```

```python
car_subdf
```

```python
# i want to applty multipile aggrigate funtions
car_df.groupby(['make','body_style'] ,as_index=False)\
    [['compression_ratio']].agg(['sum','mean'])
```

```python
#if i want to apply this on multiple colums ,we can it will prints more colums 
# i want to applty multipile aggrigate funtions
car_df.groupby(['make','body_style'] ,as_index=False)\
    [['compression_ratio','highway_mpg','price']].agg(['sum','mean'])
```

### can we perform aggrigate operations in text or non numeric or conceptual  columns
#### *Yes 👍 we can perform aggregation operations on text / non-numeric / categorical columns in pandas, but the available functions are different from numeric ones.

```python
car_df.select_dtypes(exclude='number')
```

```python
car_df.groupby(['make','body_style'] )[['num_of_cylinders']].agg(['count','max','min']).head(10)
```

## Applying diffrent .agg funtions to diffrent types of individual  colums

*  can be done usin the dictionaries

```python
car_df.groupby(['make','body_style'] )[['engine_type','num_of_cylinders','price']]\
    .agg({'engine_type':['count','max'],'num_of_cylinders':'count','price':'mean'}).
```
