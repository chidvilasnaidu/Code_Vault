## Sorting a Dataframe

    * we can sort elements in column Ascending and Descending orders
    * we can sort the multiple columns at same time 
    * We can sort multiple columns one Ascending and another Descending alternatively at the same time 

#### sort_values : used to sort the elements inside the columns
#### sort_index : used for indexes of table 

```python
import pandas as pd
car_df=pd.read_excel(r"C:\Users\chidv\Downloads\Data_sets\exported.xlsx")
car_df
```

```python
# sorting a single colunm ascending [both numeric and object]
car_df.sort_values('body_style')
```

```python
car_df.sort_values('compression_ratio')
```

```python
# sorting a single colunm desending
car_df.sort_values('make',ascending=False)
```

```python
car_df.sort_values('compression_ratio',ascending=False)
```

```python
car_df.price
```

### we can sort the multiple columns at same time 


```python
car_df.sort_values(['body_style','make']).head(15)
```

```python
#We can sort multiple columns one Ascending and another Descending alternatively at the same time 
```

```python
car_df.sort_values(['body_style','make'],ascending=[True,False]).head(15)
```

## Sorting by index

```python
car_df.sort_index(ascending=False).head(15)
```

## .reset_index()

```python
carsub=car_df.groupby(['make','aspiration'],as_index=False)[['engine_location','fuel_system','fuel_system']].count()\
        .sort_values('aspiration') # so main problem here is indexes are in UNORDERED so we wanto order it so we use reset_index()
carsub
```

```python
carsub.reset_index() #so we get  elements with orderded index and sorted aspiration
                     #but here main problem is dual index to discard it we use drop index method 
```

```python
carsub.reset_index(drop=True)
```

## TASK: Convert the MultiIndex columns into 
### ['body_style','engine_location', 'city_mpg_min', 'city_mpg_max', 'highway_mpg_mean']

```python
df_groupby = car_df.groupby(["body_style","engine_location"], as_index=False) \
[["city_mpg", "highway_mpg"]].agg(city_mpg_min=("city_mpg", "min" ),city_mpg_max=("city_mpg", "max" ),highway_mpg_mean=("highway_mpg","mean"))

df_groupby.columns
```
