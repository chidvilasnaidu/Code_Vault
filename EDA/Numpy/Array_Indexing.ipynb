# 2-D array Indexing

- arr[i] -- to get single row with index numbers i
- arr[[i,j,k...]] -- to get multiple rows with index numbers i,j,k...
- arr[i,j] -- to get the single value present in ith row and jth column
- arr[i][j] -- to get the single value present in ith row and jth column


```python
np.random.seed(52)
arr2=np.random.randint(10,150,(9,8))
arr2
```

```python
# Get the first row
arr2[0]
# Get the 3rd and 4th rows (index - 2,3)
arr2[[2,3]]
# Get the value from 1st row, 4th column
arr2[1,[4]]
arr2[1,4]
# Get the value from 4th row, 2nd column
arr2[[3,1]]
# Getthe value from 3rd row, 8th column

# Get the value from 5th row, 2th column
# Get the value from 2nd row, 4th column

```

## 3-D array Indexing 
- arr[i] -- to get enitre single 2D array with index number i
- arr[[i,j,k...]] -- to get multiple 2D arrays with index numbers i,j,k...
- arr[i,j] --  from the ith 2D array get a single row with index number j
- arr[i][j] --  from the ith 2D array get a single row with index number j
- arr[i,j,k] --  from the ith 2D array get a single integer value from jth row and kth column
- arr[i][j][k] --  from the ith 2D array get a single integer value from jth row and kth column

# Slicing

- Slicing of Arrays
- Slicing allows to extract portions of an array or select a subset of an existing array to generate new arrays. For slicing a sequence of numbers separated by colons (:) within square brackets.

`arr[row_start:row_end, col_start:col_end]`  



`Using step - arr[row_start:row_end:step, col_start:col_end:step]`

## 1-D Array Slicing 

```python
arr1d
```

```python
arr1d[0:2]
```

```python
arr1d[0:]
```

```python
arr1d[:-1]
```

```python
arr1d[-9:-1]
```

```python
arr1d[::]
```

```python
arr1d[2:8:2]
```

```python
arr1d[::-1] #reverse an array
```
