## NaN (Not a Number)

`NaN` signifies the absence of a valid value in a dataset.  
It is commonly encountered in data analysis and can arise due to several reasons:

### Causes of NaN:
- **Data entry errors**  
  e.g., Manual input mistakes that leave cells blank or incorrect.

- **Incomplete or missing data**  
  e.g., Data collected from forms or sensors may have gaps.

- **Data conversion issues**  
  e.g., Trying to convert a non-numeric string like `'abc'` into a number.

- **Mathematical operations that result in undefined values**  
  e.g., Division by zero (`0/0`) or applying math functions on string values.

> NaN values can affect calculations like mean, sum, or correlations, and are often handled using imputation, filtering, or replacement strategies in data preprocessing.

### Loading Text Data from Files

When working with **text files** that contain **tabular data**, you can load them into **DataFrames** using the correct **delimiter** that separates the values in each row.

There are several common types of delimiter-separated files:

- `\t` → **Tab-separated** files  
- `,` → **Comma-separated** (CSV) files  
- `|` → **Pipe-separated** files  
- `;` → **Semicolon-separated** files
