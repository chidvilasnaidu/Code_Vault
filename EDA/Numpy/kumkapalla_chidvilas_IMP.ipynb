# Explain the difference between a list, tuple, set, and dictionary in Python.
1. **List**
   - Ordered collection of items.
   - Allows duplicates.
   - Mutable : you can change, add, or remove elements.


2. **Tuple**
   - Ordered collection like a list.
   - Allows duplicates.
   - Immutable  once created, you cannot modify it.


3. **Set**
   - Unordered collection of unique elements.
   - Does **not** allow duplicates.
   - Mutable (you can add/remove items), but elements must be immutable (e.g., numbers, strings).


4. **Dictionary**
   - Collection of **key-value pairs**.
   - Keys must be unique and immutable.
   - Very useful for fast lookups.

````python
# Write a Python program to find the factorial of a number using recursion.
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(f"Factorial  is: {factorial(5)}")
````

````python
# Write a program to check whether a number is prime or not.
def is_prime(num):
    if num <= 1:
        return False
    
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

n = 29
if is_prime(n):
    print(f"{n} is a Prime number.")
else:
    print(f"{n} is NOT a Prime number.")
````

# What are Python generators? How are they different from iterators?

- **Generators** are special functions that use the `yield` keyword to produce values one at a time, creating an iterator automatically.  
- **Iterators** are objects with `__iter__()` and `__next__()` methods that allow iteration over elements.  
- **Difference:** Generators simplify writing iterators, use less memory (lazy evaluation), and are written like functions, while iterators are manually implemented classes/objects.

# What is NumPy and why is it used?
*NumPy** (Numerical Python) is a powerful library for numerical computing.  
It is mainly used for:

- Handling large multi-dimensional arrays and matrices.  
- Performing fast mathematical and statistical operations.  

## How is a NumPy array different from a Python list?
*Storage** → NumPy arrays are stored more compactly and use less memory, while lists are larger.  
2. **Performance** → NumPy arrays are much faster for numerical operations compared to lists.  
3. **Data Type** → NumPy arrays hold elements of the same type, whereas lists can store mixed data types.  
4. **Operations** → NumPy arrays support vectorized operations (element-wise math), but lists require loops.  
5. **Functionality** → NumPy provides advanced mathematical functions, slicing, and broadcasting, which lists do not.  

### What are the advantages of using NumPy arrays?
- NumPy arrays are faster and more memory-efficient than Python lists, support vectorized operations without loops, provide advanced mathematical functions for tasks like linear algebra and statistics, and integrate seamlessly with popular data science libraries such as Pandas, SciPy, and TensorFlow.
