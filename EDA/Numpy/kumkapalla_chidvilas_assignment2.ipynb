# NumPy Theory

## Describe the features of numpy.

- **N-Dimensional Array Object (`ndarray`)**
   * Supports multi-dimensional arrays with homogeneous data types.
   * More efficient than Python lists for large data.

- **High Performance**
  - Written in C for speed.
  - Supports fast, vectorized operations (no need for explicit loops).

- **Broadcasting**
  - Allows operations on arrays of different shapes and sizes without extra code.

- **Mathematical Functions**
  - Built-in support for math and stats functions like:
    - `sum()`, `mean()`, `std()`, `dot()`, `exp()`, `log()`, etc.

- **Random Number Generation**
  - Tools for random sampling, shuffling, and generating distributions using `numpy.random`.

- **Shape Manipulation**
  - Easily reshape, flatten, or transpose arrays:
    - `.reshape()`, `.flatten()`, `.transpose()`

- **Integration with Other Libraries**
  - Works well with popular libraries:
    - `pandas`, `matplotlib`, `scikit-learn`, `TensorFlow`, `OpenCV`, etc.

- **Linear Algebra Tools**
  - Functions for matrix operations:
    - Inverse, determinant, eigenvalues via `numpy.linalg`.

- **Handling Missing Data**
  - Supports `NaN` and masked arrays to manage missing or invalid values.

- **Memory Efficiency**
  - Uses contiguous memory blocks.
  - Consumes less memory and accesses data faster than native Python structures.


### Write the applications of numpy.
* Ideal for solving large-scale mathematical problems
* Used for handling and processing large datasets.
* Financial and Statistical Analysis
* Essential for working with input features, weight matrices, and datasets.
* Supports efficient vectorized operations used in ML libraries like TensorFlow and scikit-learn.

## How numpy is different from python sequences.
1. **Performance & Efficiency**  
   - NumPy Arrays: Faster due to C backend and contiguous memory.  
   - Python Lists: Slower due to dynamic typing and interpreter overhead.  

2. **Data Homogeneity**  
   - NumPy Arrays: Store elements of the **same type** (homogeneous).  
   - Python Lists: Can store **mixed types** (heterogeneous).  

3. **Size Mutability**  
   - NumPy Arrays: **Fixed size** (resizing creates a new array).  
   - Python Lists: **Dynamic size** (can grow/shrink with `append()`, `pop()`).  


4. **Multidimensional Support**  
   - NumPy Arrays: Native **n-dimensional arrays** (e.g., matrices).  
   - Python Lists: Need nested lists (less efficient for math ops).  
  

5. **Memory Usage**  
   - NumPy Arrays: **Memory-efficient** (compact storage).  
   - Python Lists: Higher memory overhead (stores extra metadata).

```python
#Create a NumPy array of size 6. 
#Replace the second element with 99. 
#Update a slice of elements from index 2 to 4 with 0. Set the last three elements to 10.
arr6=np.random.randint(1,20,6)
np.random.seed(45)
print('Create a NumPy array of size 6 : \t\t \t:',arr6)
arr6[1]=99
print('after Replace the second element with 99. \t\t:',arr6)
arr6[2:4]=0
print('Update a slice of elements from index 2 to 4 with 0 \t:',arr6)
arr6[-3:]=10
print('Set the last three elements to 10.\t\t \t:',arr6)
```

```python
# Create a 3x3 NumPy array with values from 1 to 9. Replace the second row with all 5s. Update the third column with values [8, 9, 10].
arr39=np.arange(1,10).reshape(3,3)
print(arr39)
arr39[2]=5
print('after replacing second row with 5 \n',arr39)
```

```python
#Create two arrays: A = np.array([2, 4, 6]), B = np.array([1, 2, 3]). Perform element-wise addition, subtraction, multiplication, and division
A = np.array([2, 4, 6])
B = np.array([1, 2, 3])
print("Addition :",A+B)
print("subtraction :",A-B)
print("multiplication :",A*B)
print("division :",A//B)
```

```python
#Create two 2×2 matrices and compute their dot product using np.dot() and @ operator.
A = np.random.randint(10,58,(3,3))
B = np.random.randint(2,9,(3,3))
res=np.dot(A,B)
print('dot product result ..\n',res)
print('@ product result ..\n',A@B)
```

```python
#create an array with values [4, 8, 15, 16, 23, 42]. Compute sum, min, max, mean, median, variance, standard deviation, and correlation coefficient.
arr=np.array([4, 8, 15, 16, 23, 42])
print('sum is :',sum(arr))
print('min is :',min(arr))
print('max is :',max(arr))
print('mean is :',np.mean(arr))
print('median is :',np.median(arr))
print('variance is :',np.var(arr))
print('standard deviation :',np.corrcoef(arr))
print('correlation coefficient. is :',np.std(arr))
```
