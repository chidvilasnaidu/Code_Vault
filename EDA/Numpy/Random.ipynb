## Numpy Random Numbers

* The np.random.rand method, generates an array with random numbers that are uniformly distributed between 0 and 1.
* The np.random.randn method, generates an array with random numbers that are normally distributed with mean = 0 and sd = 1.
* The np.random.randint method, generates an array with random numbers that are uniformly distributed between 0 and given integer.
* The np.random.uniform method, generates random float values.



## rand

```python
import random 
import numpy as np
print(f" generates a random number  b/w 0,1  :  {np.random.rand() }")
print(f" \n generates a random 1-D array  b/w 0,1  :  {np.random.rand(3) }")
print(f" \n generates a random 2-D array  b/w 0,1 by given shape  :  \n\n{np.random.rand(4,4)}")
arr=np.random.rand(4,7)
```

# randn -
* The np.random.randn method, generates an array with random numbers that are normally distributed with mean = 0 and sd = 1

```python
import random 
import numpy as np
print(f" generates a random number  b/w 0,1  :  {np.random.randn() }")
print(f" \n generates a random 1-D array  b/w 0,1  :  {np.random.randn(3) }")
print(f" \n generates a random 2-D array  b/w 0,1 by given shape  :  \n\n{np.random.randn(3,4,4)}")
arr=np.random.rand(4,7)
```

# random.randint
* •	The np.random.randint method, generates an array with random numbers that are uniformly distributed between 0 and given integer.

```python
import random 
import numpy as np
print(f" \n generates a random number  b/w 0,1  :  {np.random.randint(5) }") # if only one number is given then min==0, max =given integer 
print(f"\n  generates a random number  b/w 0,1  :  {np.random.randint(2,5) }")
print(f" \n generates a random 1-D array  b/w 0,1  :  {np.random.randint(10,150,6) }")
print(f" \n generates a random 2-D array  b/w 0,1 by given shape  :  \n\n{np.random.randint(10,150,(4,5))}")
arr=np.random.rand(4,7)
```

# np.random.uniform  
*  The np.random.uniform method, generates random float values.m

```python
import random 
import numpy as np
print(f"\n Generates a random float number  b/w 0,1  :  {np.random.uniform(2,5) }")
print(f" \n generates a random float 1-D array  b/w 0,1  :  \n{np.random.uniform(10,150,6) }")
print(f" \n generates a random float 2-D array  b/w 0,1 by given shape  :  \n\n{np.random.uniform(10,150,(4,5))}")
arr=np.random.rand(4,7)
```
