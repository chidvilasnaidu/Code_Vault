## ii. Arrays with linspace()
*The linspace() function generates an array with evenly spaced values between specified start and end values, using a specified number of elements.*

Syntax:

np.linspace(Start, End, Number of elements)

```python
# so we can make or create the evenly seperated elements by
arr2=np.linspace(1,10,10) # so it will evenly divides the 10 numbers by 1
print(arr2) 
print('--'*30)
print()
arr2=np.linspace(1,10,15) # so it will evenly divides the 10 numbers by 15
print(arr2) 
print('--'*30)
print()
```

```python
arr2=np.linspace(1,10,20) # so it will evenly divides the 10 numbers by 20
print(arr2) 
print('--'*30)
arr2=np.linspace(1,10,20) # so it will evenly divides the 1,11  numbers by 25 2d elements 
arr2d=np.linspace(1,11,25).reshape(5,5)
print(arr2d) 
print('--'*30)
```

## vii. Eye Array
The eye() function, returns an array where all elements are equal to zero, except for the k-th diagonal, whose values are equal to one.

```python
arri=np.eye(4,k=0)
print("defauly eye array is :\n",arri) #4,k=0
```

```python
arri=np.eye(5,k=2)
print("5,k=2 :\n",arri) #4,k=0
```

```python
arri=np.eye(5,k=-2)
print("5,k=-2 :\n",arri) #4,k=0
```

```python
arri=np.eye(7,8,k=-1)
print("user defined shape :\n",arri) #4,k=0
```
