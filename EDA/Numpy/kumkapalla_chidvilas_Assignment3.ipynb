```python
#16. Create a 3x3 array and apply the following operations element-wise: square, log, exponential, and sine
import numpy as np

arr = np.random.randint(1, 10, (3, 3))
print("Original Array:\n", arr)

print("\nsquared:\n", np.square(arr))
print("\nlog:\n", np.log(arr))
print("\nExponential:\n", np.exp(arr))
print("\nSine:\n", np.sin(arr))
```

```python
#Create a 4x4 matrix with random integers between 1 and 100. Normalize the matrix using broadcasting
arr = np.random.randint(1, 101,(4, 4))
print("Original Matrix:\n", arr)
norm = (arr - arr.min()) / (arr.max() - arr.min())
print("\nNormalized Matrix:\n", norm)
```

```python
#18. Create an array of 5 angles in degrees. Convert to radians and calculate sin and cos using np.pi.
angles_deg = np.array([0, 30, 45, 60, 90])
angles_rad = angles_deg * (np.pi / 180)

print("Angles in Degrees:\n", angles_deg)
print("\nAngles in Radians:\n", angles_rad)
print("\nSine Values:\n", np.sin(angles_rad))
print("\nCosine Values:\n", np.cos(angles_rad))
```

```python
##19. Given an array with float values, convert it to integer using astype and calculate the sum before and after conversion.
arr = np.array([1.7, 2.5, 3.9, 4.1, 5.8])
print("Original Float Array:\n", arr)

sum_before = np.sum(arr)
arr_int = arr.astype(int)
sum_after = np.sum(arr_int)

print("\nSum Before Conversion:", sum_before)
print("Converted Integer Array:\n", arr_int)
print("Sum After Conversion:", sum_after)
```

```python
#25. Use np.where to find positions of all elements divisible by 3 in a randomly generated 5x5 matrix.

arr = np.random.randint(1, 101, (5, 5))
print("original array:\n", arr)

positions = np.where(arr % 3 == 0)
print("\npositions of elements divisible by 3:\n", list(zip(positions[0], positions[1])))

```

```python
##26. Create a 2D array of shape (6,4). Find:

arr = np.random.randint(1, 101, (6, 4))
print("original array:\n", arr)

mean_cols = np.mean(arr, axis=0)
std_all = np.std(arr)
median_rows = np.median(arr, axis=1)

print("\nmean of each column:", mean_cols)
print("standard deviation of entire array:", std_all)
print("median of each row:", median_rows)
```

```python
#27. Generate a random array of 1000 values using randn. Count how many values fall between -1 and 1 (hint: normal distribution rule).
arr = np.random.randn(1000)
print("generated array:\n", arr)

count = np.sum((arr > -1) & (arr < 1))
print("\nnumber of values between -1 and 1:", count)
```

```python
#28. Create a 5x5 array and calculate row-wise and column-wise sum and mean.

arr1 = np.random.randint(1, 99, (5, 5))
print("array is:\n", arr1)
row_sum = np.sum(arr1, axis=1)
row_mean = np.mean(arr1, axis=1)
col_sum = np.sum(arr1, axis=0)
col_mean = np.mean(arr1, axis=0)
print("\nrow-wise sum:", row_sum)
print("row-wise mean:", row_mean)
print("\ncolumn-wise sum:", col_sum)
print("column-wise mean:", col_mean)
```
