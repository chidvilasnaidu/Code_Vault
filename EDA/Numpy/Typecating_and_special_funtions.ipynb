## Flatten and Ravel

`flatten():`

Always returns a copy of the original array. This means a new, independent array is created in memory, and any modifications made to the flattened array will not affect the original array.

`ravel():`

Returns a view of the original array whenever possible. A view is essentially a reference to the original data, meaning they share the same memory.

```python
flattened=arr2d.flatten()
raveled=arr2d.ravel()
```

```python
flattened 
arr2d
```

```python
raveled
arr2d
```

# 🧩 Splitting of Arrays in NumPy

Splitting means dividing one array into multiple sub-arrays.

### 📌 Types of Splitting
- **`np.split()`** → Divides an array into equal parts along a given axis.  
- **`np.hsplit()`** → Splits horizontally (column-wise).  
- **`np.vsplit()`** → Splits vertically (row-wise).  
- **`np.dsplit()`** → Splits along the depth (3rd axis).  

✅ Use when you want to break arrays into smaller chunks for processing.


```python
bigarr=np.random.randint(1,100,(8,8))
bigarr
```

# # 🔑 Rule for Splitting Arrays

If you want to split an array into *n* parts,  
then *n* must be a **factor of the array size along that axis**.  

👉 Example:  
- Array of size 6 → can be split into 2, 3, or 6 equal parts.  
- Array of size 8 → can be split into 2, 4, or 8 equal parts.  
- If the split number is **not a factor**, NumPy will throw an error.


```python
bigarr.shape
```

```python
np.split(bigarr,2) # here 2 is  array will be divided into N equal arrays 
```

```python
np.vsplit(bigarr,4) #vertical wise divided into 4 diff partsAa
```

```python
np.hsplit(bigarr,4)
```
