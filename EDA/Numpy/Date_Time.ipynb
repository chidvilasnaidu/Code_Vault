##  Cheat code 'date' table 
- %d -------> Represents Day in number 
- %m -------> Represents months in number
- %Y -------> Represents year in full ex :2025
- %y -------> Represents year in Short ex :25
- %B -------> Represents Full Month name 
- %b -------> Represents Short Month name
- %a -------> Represents Full Day name
- %A -------> Represents Short Day name


##  Cheat code 'Time ' table 
- %H -------> Represents hours in 12 hr format 
- %I -------> Represents hours in 24 hr format 
- %p -------> Represents AM/PM
- %M -------> Represents Minute 
- %S -------> Represents Seconds `

## `freq `
- parameter inside .to_period(freq) tells what time span (granularity) the period should represent.

| Freq Code | Description                          | Example Output (for 2024-09-02) |
|-----------|--------------------------------------|---------------------------------|
| D         | Daily                                | 2024-09-02                      |
| W         | Weekly (ends on Sunday by default)   | 2024-08-26/2024-09-01           |
| M         | Month End                            | 2024-09                         |
| MS        | Month Start                          | 2024-09                         |
| Q         | Quarter End (default: Dec year-end)  | 2024Q3                          |
| Q-DEC     | Quarter ending in December           | 2024Q3                          |
| Q-MAR     | Quarter ending in March              | 2024Q2                          |
| A or Y    | Year End (default: Dec)              | 2024                            |
| AS        | Year Start                           | 2024                            |
| H         | Hourly                               | 2024-09-02 00:00                |
| T or min  | Minutely                             | 2024-09-02 00:00                |
| S         | Secondly                             | 2024-09-02 00:00:00             |

# Date Range 

### Pandas has a super useful function pd.date_range() for creating a sequence of datetimes.

### pd.date_range you can’t use periods and freq with both start and end at the same time.
Rules

If you use start + end + freq → Pandas figures out all the timestamps between them.

If you use start + periods + freq → Pandas generates n periods forward.

If you use end + periods + freq → Pandas generates n periods backward.

You cannot mix start, end, periods all together.

```python
pd.date_range(start='2019-03-06', end='2019-03-12')
```

```python
pd.date_range(start='2019-03-06 00:00:00', end=' 2019-04-08 23:00:00')
```

```python
pd.date_range(start='2019-03-06 00:00:00', end=' 2019-04-08 23:00:00',periods=10) # it is like Bins automatecally shows only 10 values
```

```python
pd.date_range(start='2019-03-06 00:00:00',periods=10,freq='MS') #Month starting 
```

```python
pd.date_range(start='2019-03-06 00:00:00',periods=10,freq='Q') #Quater 
```

```python
pd.date_range(start='2019-03-06 00:00:00',periods=10,freq='M') #monthend
```

```python
pd.date_range(start='2019-03-06 00:00:00',periods=10,freq='W') #weekly
```

```python
pd.date_range(start='2019-03-06 00:00:00',periods=10,freq='B') # Only Business days 
```

```python
pd.date_range(start='2019-03-06 00:00:00',periods=10,freq='H') #hourly
```

```python
pd.date_range(start='2019-03-06 00:00:00',periods=10,freq='D ') #daily
```
