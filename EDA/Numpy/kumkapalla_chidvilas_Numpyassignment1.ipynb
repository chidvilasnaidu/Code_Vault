## Reshape

```python
#Write code to reshape a 1D array of 12 elements into a 3x4 array.
import random as rd
arr12=rd.sample(range(10,150),12)
print(np.reshape(arr12,(4,3)))
```

```python
#What will happen if you try to reshape an array with incompatible dimensions?
# it will throw an errors
# -> arrays can be only reshaped if shape is the factors of size

print(np.reshape(arr12,(4,3,2))) #  <<---------- Throws the error
```

```python
#Convert a 2x6 array into a 3x4 array using reshape.
arr12=np.array(rd.sample(range(1250,1500),12)).reshape(2,6)
print("2x6 Array : \n", arr12)
print("3x4 after reshaping Array : \n", arr12.reshape(3,4))
```

```python
#How to reshape a 3x3 array into a 1D array?
import random as rd
lst=rd.sample(range(1500,1980),9)
arr2d=np.array(lst)
arr2d=arr2d.reshape(3,3)
print("3x3 Array : \n", arr2d)
print("\n\n 3x3 after reshaping into 1d  Array : \n\n", arr2d.reshape(9,))
```

```python
#What is the use of -1 in reshape function?
# reshape(1,-1)----> it will reshapes the array  1 row x n columns
#reshape(-1,1) ----> it will reshapes the array to n rows x 1 column

lst=[range(1,10)]
arr2d=np.array(lst)
print('reshape(1,-1) of matrix gives',arr2d.reshape(1,-1))
print('reshape(-1,1) of matrix gives',arr2d.reshape(-1,1))
```

## Resize

```python
#Differentiate between reshape and resize in NumPy.
#	reshape() returns a new array with a different shape without modifying the original array.
#	resize() modifies the original array and can also change its size by adding or removing elements.

arr_reshape=np.arange(1,101,9)
arr_resize=np.arange(1,101,9)
print('reshaping the array : \n',arr_reshape.reshape(4,3))
print('Original array after reshape : \n',arr_reshape)  #it doesnt modifies the original 1-D array
print()
print('resize the array : \n',arr_resize.resize(4,3))
print('Original array after resize : \n',arr_resize)  #it doesnt modifies the original 1-D array
```

## Write code to resize an array in-place.
* resize() modifies the original array in-place.
* To resize an array in-place, you must ensure it's a base array (not a view). 
* It works only if the array is not a view (e.g., don't use .reshape() directly before resize()).

## What is the difference between np.eye and np.identity?
* np.identity(n) creates a square identity matrix only.

* np.eye(N, M, k) is more flexible — allows non-square matrices and shifting the diagonal using k.

## fliplr, flipud, flip, rot90

```python
#Write code to flip a 2D array left to right.
arr=np.array(rd.sample(range(0,99),9)).reshape(3,3)
print(f"original arr is :\n",arr)
flipped_lr = np.fliplr(arr)
print(f"flipped  arr is :\n",flipped_lr)
```

```python
#Flip a 3x3 matrix upside down.
flipped_ud = np.flipud(arr)
print(flipped_ud)
```

```python
#How to reverse all rows in a 2D array?
reversed_rows = arr[:, ::-1]
print(reversed_rows)
```

```python
#Write code to rotate a 2D array 90 degrees counter-clockwise.
rotated_ccw = np.rot90(arr)
print(rotated_ccw)
```

```python
#Explain the use of np.flip and how it differs from flipud or fliplr.
np.flip(arr)         
np.flip(arr, axis=0)  
np.flip(arr, axis=1)
```
