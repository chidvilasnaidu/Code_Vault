```python
#dimonf
n=5
for i in range(1,n,2):
    sp=(n-i)//2
    print(' '*sp+'*'*i)
for j in range(n,0,-2):
    sp2=(n-j)//2
    print(' '*sp2+'*'*j)
```

```python
n=5
for i in range(1,n,2):
    sp=(n-i)//2
    print(' '*sp,'*'*i)
for j in range(n,0,-2):
    sp2=(n-j)//2
    print(' '*sp2,'*'*j)
```

```python
n=4
for i in range(1,n+1):
    print(" "*(n-i),end=' ')
    for j in range(1,i+1):
        print(i,end =' ')
    print()
for j in range(n-1,0,-1):
    print(" "*(n-j),end=' ')
    for j in range(1,j+1):
        print(j,end =' ')
    print()

```

```python
# symmetric pyramid.
#hill pattren
#same thing we are using with numbers

n=5
for i in range(1,n+1):
    left_spaces='  '*(n-i)# we must use the double space to reach goal 
    left_stars='* '*i
    right_stars='* '*(i-1)
    print(left_spaces+left_stars+right_stars)
```

```python
#pyramid approach 2
n=5
for i in range(1,n+1):
    print('  '*(n-i),'* '*i+'* '*(i-1))
```

```python
n=5
for i in range(1,n+1):
    print('  '*(n-i),end=' ')
    for j in range(1,i+1):
        print(i,end=' ')
    for j in range(i-1,0,-1): #we simply decrement the values
        print(i,end=' ')
    print()
     
```
