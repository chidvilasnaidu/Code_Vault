## Grey Video

```python
import cv2
capture =cv2.VideoCapture(r"D:\DS\Notes\CV\plant.mp4")
while True:
    istrue, fig=capture.read()
    if istrue:
        grey=cv2.cvtColor(fig, cv2.COLOR_BGR2GRAY)
        Resize=cv2.resize(grey,(650,450))
        cv2.imshow('Play', Resize)
        key=cv2.waitKey(40)
        if(key== ord('q')):
            break
    else:
        break
cv2.destroyAllWindows()
```

## Apling Blur on video







```python
import cv2
cap=cv2.VideoCapture(r"D:\DS\Notes\CV\plant.mp4")
while True:
    istrue, fig=cap.read()
    if istrue:
        Resize=cv2.resize(fig,(650,450))
        Blurred=cv2.blur(Resize,(3,3))
        cv2.imshow('cap',Blurred)
        
        key=cv2.waitKey(40)
        if key==ord('q'):
            break
    else:
        break 
cv2.destroyAllWindows()
```

## Edge Detection 

```python
import cv2
capture =cv2.VideoCapture(r"D:\DS\Notes\CV\plant.mp4")
while True:
    istrue, fig=capture.read()
    if istrue:
        Resize=cv2.resize(fig,(650,450))
        Out=cv2.Canny(Resize,50,150)
        cv2.imshow('Play', Out)
        key=cv2.waitKey(40)
        if(key== ord('q')):
            break
    else:
        break
cv2.destroyAllWindows()
```
