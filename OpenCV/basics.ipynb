#### Bitwise Operations 

```python
# AND OR XOR 
# Same like T, F but here instead of values it will use Pixels as Operandas 
# For Clear Understand we are using User Defined Colour Intensities 
arr1=np.ones((400,400), dtype=np.uint8)
arr1[100:250, 100:250]=255
cv2.imshow('a1',arr1)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
arr2=np.ones((400,400), dtype=np.uint8)
arr2[150:300, 150:300]=255
cv2.imshow('a1',arr2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## Now Both arr1, arr2 are In diffrent positions so we perform Bitwise and 
## AND 
#0 1 = 0   0 --> Black pix , 1 --> white pix 
#1 0 = 0
#0 0 = 0
#1 1 = 1
out= cv2.bitwise_and(arr1,arr2)
cv2.imshow('Image 1',arr1)
cv2.imshow('Image 2',arr2)
cv2.imshow('AND b/w img 1&2 ',out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

## OR  
#0 1 = 1   0 --> Black pix , 1 --> white pix 
#1 0 = 1
#0 0 = 0
#1 1 = 1
out= cv2.bitwise_or(arr1,arr2)
cv2.imshow('Image 1',arr1)
cv2.imshow('Image 2',arr2)
cv2.imshow('OR b/w img 1&2 ',out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

## XOR  
#0 1 = 1   0 --> Black pix , 1 --> white pix 
#1 0 = 1
#0 0 = 0
#1 1 = 0
out= cv2.bitwise_xor(arr1,arr2)
cv2.imshow('Image 1',arr1)
cv2.imshow('Image 2',arr2)
cv2.imshow('XOR b/w img 1&2 ',out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

mask = np.zeros((500, 500, 3), dtype=np.uint8)

# Points
points = np.array([[
    [150, 50],
    [250, 50],
    [300, 150],
    [250, 250],
    [150, 250],
    [100, 150]
]], dtype=np.int32)

# Draw polygon
cv2.polylines(mask, points, True, (255, 255, 255), 2)

for point in points[0]:
    x, y = point
    cv2.circle(mask, (x, y), 5, (0,0,255), -1)
    cv2.putText(mask, f"({x},{y})", (x+5, y-5),
                cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,255,0), 1)

# Show image
cv2.imshow('Points + Polygon', mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 2.  Gaussian Blur:  Sigma (σ) in Gaussian Blur represents the standard deviation and controls the amount of smoothing by determining how much neighboring pixels contribute to the blur.

```python
img=cv2.imread(r"D:\DS\Notes\CV\GOLD.jpg")
grey= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
Output=cv2.GaussianBlur(grey, (3,3), sigmaX=60)
cv2.imshow('Clurred  ',Output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
