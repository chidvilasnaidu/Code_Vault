```python
## filehandeling involvers in the several modes
### Write: write mode which is used to create and write data into files ,mode='w'
### append : It is also used for writing the data into file,      mode ='a'
### Read:used to read or load the data :
            further divided into three type:
                    read: read the entire data
                    readline: read only first line
                    readlines: Returns the list of all sentences or lines
```

```python
### Write: write mode which is used to create and write data into files
with open("C:/Users/chidv/OneDrive/Desktop/Innomatics/Inno_py/Notes/hello.txt" ,'w') as file: 
    #if hello.txt is is not exixting it will create new file 
    # if it already existing it will override or replace tex with new one  
    file.write("Hello chiddu How are you")
    file.close()
```

```python
# append : It is also used for writing the data into file
with open("C:/Users/chidv/OneDrive/Desktop/Innomatics/Inno_py/Notes/hello.txt" ,'a') as file: 
    file.write("\n iam back, comming from the append section")
    file.close()
```

```python
# Read:used to read or load the data :
with open("C:/Users/chidv/OneDrive/Desktop/Innomatics/Inno_py/Notes/hello.txt" ,'r') as file: 
    print(file.read())
    file.close()
```

```python
#readline: only returns the first line
with open("C:/Users/chidv/OneDrive/Desktop/Innomatics/Inno_py/Notes/hello.txt" ,'r') as file: 
    print(file.readline())
    file.close()
```

```python
#readlines: it the list of sentences 
with open("C:/Users/chidv/OneDrive/Desktop/Innomatics/Inno_py/Notes/hello.txt" ,'r') as file: 
    print(file.readlines())
    file.close()
```

```python
def prime(num):
    flag=False
    if(num==0 or num==1 ):
        flag=True
    else:
        for i in range(2,num):
            if(num%i==0):
                flag=True
            else:
                pass
    if flag:
        print('non prime')
    else:
        print("prime")
```

```python
prime(4)
```
