### Write a Python script to collect 10 cat facts (each ≤ 80 characters) from the public API https://catfact.ninja/fact. Store them with unique IDs (fact1, fact2, ...) in a dictionary, then convert the data into a pandas DataFrame with two columns: id and facts, and display both.

```python
import requests as rd
facts={}
lst1=[]
lst2=[]
facts_dir={}
data_pnt=1
while data_pnt<=10:
    v1=rd.get("https://catfact.ninja/fact")
    data=v1.json()
    if(data['length']<=80):
        facts[f"fact{data_pnt}"]=data['fact']
        data_pnt+=1
    else:
        pass
    
for keys,values in facts.items():
    lst1.append(keys)
    lst2.append(facts[keys])
facts_dir['id']=lst1
facts_dir['facts']=lst2
print(facts_dir)   
```

```python
import pandas as pd
import requests as rq
database={'name':[],'gender':[],'country':[],'state':[],'city':[], 'Streetname':[],'Email':[],'username':[],'password':[],'phone':[]}
for i in range(1,11):
    temp=rq.get('https://randomuser.me/api/')
    var=temp.json()
    person = {
        'name': var['results'][0]['name']['first'] + " " + var['results'][0]['name']['last'],
        'gender': var['results'][0]['gender'],
        'country': var['results'][0]['location']['country'],
        'state': var['results'][0]['location']['state'],
        'city': var['results'][0]['location']['city'],
        'Streetname': var['results'][0]['location']['street']['name'],
        'Email': var['results'][0]['email'],
        'username': var['results'][0]['login']['username'],
        'password': var['results'][0]['login']['password'],
        'phone': var['results'][0]['phone']
    }
    for every in database:
        database[every].append(person[every])

pd.DataFrame(database)
```
