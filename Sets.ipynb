```python
#union 
set1={8,7,9,12}
set2={1,9,11}
set1.union(set2)
```

```python
#difference
set1={1,2,6,9,12}
set2={1,2,3,6}
print(set1.difference(set2))
print(set1)
```

```python
#difference_update
set1={1,2,6,9,12}
set2={1,2,3,6}
set1.difference_update(set2)
print(set1)
```

```python
# add
set2={1,2,3,6}
set2.add(25)
print(set2)
```

```python
#discard
set1={1,2,6,9,12}
set1.discard(45)
print(set1)
```

```python
#remove
set1={1,2,6,9,12}
set1.remove(45)
print(set1)
```

```python
#Intersection
set1={1,2,6,9,12}
set2={1,2,3,6}
print(set1.intersection(set2))
print(set1)
```

```python
#intersection_update
set1={1,2,6,9,12}
set2={1,2,3,6}
print(set1.intersection_update(set2))
print(set1)
```

```python
#update
set1={1,2,6,9,12}
set2={1,2,3,6}
set1.update(set2)
print(set1)
```
