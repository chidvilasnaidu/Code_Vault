## Access Modifiers 
## public private protected

```python
class cls1:
    x1 = 25         # Public member
    __x2 = 15       # Private member
    _x3 = 10        # Protected member

print('\nAccessing public members with class and object\n')
print('\tCalling by class:', cls1.x1)   # ✅ Public access by class
obj = cls1()
print('\tCalling by object:', obj.x1)   # ✅ Public access by object

print('\nAccessing private members with class and object (using name mangling)\n')

print('\tCalling by class:', cls1._cls1__x2)  # ✅ Correct (name mangling)
print('\tCalling by object:', obj._cls1__x2)  # ✅ Correct

print('\nAccessing protected members with class and object\n')
print('\tCalling by class:', cls1._x3)        # ✅ Works, but discouraged outside class
print('\tCalling by object:', obj._x3)        # ✅ Works, but discouraged
```

```python
#protected : protected
# Accessible within the class and its subclasses.
# Syntax (Python): Prefix with a single underscore _.
# ⚠️ It's a convention, not strict protection in Python.
class Engine:
    _rpm=550  # protected varible  
    value=550 
    def _Details():
        return f'This enging belongs to the Varun brand'
class motor(Engine):  # subclass
    pass
#accessing the protected ones   classname_var like obj._var or method
print('\t Accessing via class ')
print('Accessing the protected var :', Engine._rpm)
print('Accessing the protected method by class:', Engine._Details())
obj=Engine
print('Accessing the protected method by object :', obj._Details())
ob2=motor
print('\n\t Accessing via Subclass ')
print('Accessing the protected var :', motor._rpm)
print('Accessing the protected method by subclass:', motor._Details())
obj=Engine
print('Accessing the protected method by subclass object :', ob2._Details())
```
