```python
from langchain_core.prompts import HumanMessagePromptTemplate, SystemMessagePromptTemplate, ChatPromptTemplate,
```

```python
user_temp="Hello Bot, Can you please able to create {topic} Web application, Using HTML,CSS,JS"
system_msg= ''' Hello, system Behave like a experienced Developer ,You need to produce the codes based on human prompt, you must crosscheck before producing the code
if  the User ask About python give like

!!!! Make sure you did not give any english words or sentenceces as a reply I only start the code with below format
NO  "",  `'',

START EMPTY : LIKE
def fun():
  jkjjkf

END With 5 dashes ----- NO python headingss , no etc




if he ask about Web Page or Application


 code.... even dondt give HTML heading
-----
Likewise CSS code

 code....even dondt give css heading
-----
LikeWise JAVASCRIPT code

 code....even dondt give JS heading
-----
NO SINGLE WORD XTRA THAN CODE AND SEPERATORS
DONT GIVE HEDING ABOVE CODE LIKE X html X
MAke sure each file is interlinked with names of "index.html""style.css""script.js
if html was css you link the css file with style.css name likewise script.js
"
'''
```

```python
user_msg= HumanMessagePromptTemplate.from_template(user_temp)
Prompt_var= ChatPromptTemplate.from_messages([system_msg, user_msg])
```

```python
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
import os
key= userdata.get('GEMINI3')
os.environ['GOOGLE_API_KEY']=key
```

```python
Model=ChatGoogleGenerativeAI(
    model="gemini-2.5-flash")
Model
```

```python
from langchain_core.output_parsers import StrOutputParser
parser=StrOutputParser()
parser
```

```python
# LECL Chain

chain=Prompt_var|Model|parser
response=chain.invoke({
    "topic":"Relastic Clock with stop watch, timer, timeZones  "
})
```

```python
print(response)
```

```python
html=response.split(sep='-----')[0]
css=response.split(sep='-----')[1]
js=response.split(sep='-----')[2]
```

```python
with open(r'index.html', 'w') as f:
  f.write(html)
with open(r'style.css', 'w') as f:
  f.write(css)
with open(r'script.js', 'w') as f:
  f.write(js)
```
