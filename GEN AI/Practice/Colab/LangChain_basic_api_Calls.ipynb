## Groq

```python
pip install -qU langchain-groq
```

```python
from google.colab import userdata
import os
Gork_Key= userdata.get('Gork_Key')
os.environ['GROQ_API_KEY']=Gork_Key
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="qwen/qwen3-32b",
)
```

```python
messages = [
    (
        "system",
        "you are a Polite student you will explain every thing min  2-3 lines even greetings ",
    ),
    ("human", "Hello Bot what is your name "),
]
ai_msg = llm.invoke(messages)
ai_msg.text
```

### What is the main diffrence between Giving seperate Model function calls in Langchain Vs Using Init Chat model?
- Using separate model classes gives fine-grained control and access to provider-specific features, making it suitable for production and complex applications. `init_chat_model()` provides a unified interface across different providers, simplifying model switching and rapid


## init_chat_model()

```python
from langchain.chat_models import init_chat_model
```

```python
from langchain_groq import ChatGroq
```

```python
model=init_chat_model(
    model='qwen/qwen3-32b',
    model_provider='groq'
)
```

```python
messages = [
    (
        "system",
        "You are a helpful assistant that translates English to Hindi. Translate the user sentence.",
    ),
    ("human", "Hello Good morning Iam Kumkapalla Chidvilas , An recent graduate from NITJ in information Technology. I am orignally from Gudivada AP, And I have good knowlede in DS &AI , I completed multiple projects on RAG System, Ihave indepth knowledge in GEN AI. I love french people and culture "),
]
msg = model.invoke(messages)
msg.text
```
