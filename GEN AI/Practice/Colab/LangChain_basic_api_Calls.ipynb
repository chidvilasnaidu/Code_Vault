```python
from google.colab import userdata
import os
Gem_key= userdata.get('Gemini')
```

```python
os.environ['GOOGLE_API_KEY']=Gem_key
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",

)
```

```python
messages = [
            ("system"," Respond in Polite and welcoming manner, with good morning "),
            ("user", "Hello!")]

response = model.invoke(messages)
```

```python
response.text
```

## Groq

```python
pip install -qU langchain-groq
```

```python
from google.colab import userdata
import os
Gork_Key= userdata.get('Gork_Key')
os.environ['GROQ_API_KEY']=Gork_Key
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="qwen/qwen3-32b",
)
```

```python
messages = [
    (
        "system",
        "you are a Polite student you will explain every thing min  2-3 lines even greetings ",
    ),
    ("human", "Hello Bot what is your name "),
]
ai_msg = llm.invoke(messages)
ai_msg.text
```

```python
from google.colab import userdata
import os
Mistral_api= userdata.get('Mistral_api')
os.environ['MISTRAL_API_KEY']=Mistral_api
```

```python
from langchain_mistralai import ChatMistralAI

llm = ChatMistralAI(
    model="mistral-large-latest",
)
```
