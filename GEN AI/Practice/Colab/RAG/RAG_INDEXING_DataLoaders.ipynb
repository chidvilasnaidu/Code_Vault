# RAG: Retrival Agumented Generation
RAG is an AI architecture that combines an LLM with a retrieval system. Instead of relying solely on training data, RAG first retrieves the most relevant information from an external knowledge source  (like a vector database), then feeds them as context to the LLM. This makes responses more accurate, up-to-date, and grounded in real information. It's widely used in enterprise chatbots and Q&A systems

RAG consists of 3 Phases:


1.   INDEXING
2.   Retrival
3.   Generation




`Indexing `


*   loading documents
*   text spliting
*   creating embbedings
*   Storage

Creating Small system like RAG

```python
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate, ChatPromptTemplate
```

```python
systemMsg=SystemMessagePromptTemplate.from_template('Just read given Document very carefully, understand document, Understand logically answer to user qurey carefully without any halloginations, user will mostly ask qsn from below text so give it in SINGLE SENTENCE , so content Text is {content}')
HumanMsg=HumanMessagePromptTemplate.from_template('  {letter} ')
```

```python
PromptVar=ChatPromptTemplate.from_messages([systemMsg,HumanMsg])
```

```python
pip install -U langchain-google-genai
```

```python
from google.colab import  userdata
import os
from langchain_google_genai import ChatGoogleGenerativeAI
os.environ['GOOGLE_API_KEY']=userdata.get('GEMINI3')
model = ChatGoogleGenerativeAI(
     model="gemini-2.5-flash")
```

```python
from langchain_core.output_parsers import  StrOutputParser
parser=StrOutputParser()
```

```python
chain =PromptVar|model|parser
chain
```

```python
# We extract the text using pdf text loaders
from langchain_opendataloader_pdf import OpenDataLoaderPDFLoader

loader = OpenDataLoaderPDFLoader(
    file_path=[r"/content/Motivation Letter.pdf"],
    format="text"
)
documents = loader.load()
```

```python
Text=documents[0].page_content
Text
```

```python
res=chain.invoke({
    'content':Text,
    'letter':'what collage did the student is applying for '
})
```
