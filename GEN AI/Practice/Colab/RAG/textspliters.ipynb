```python
# Loading the data using data loaders
# text loaders
from langchain_community.document_loaders import TextLoader
loader_var=TextLoader(r'/content/DL_Questions.txt')
textresult=loader_var.load()
Text=textresult[0].page_content
```

**CharacterTextSplitter** : Here we need to provide the seperators manually inorder split the text

```python
from langchain_text_splitters import CharacterTextSplitter
```

**CharacterTextSplitter** : We must give the type of a Seperator
**RecursiveCharacterTextSplitter**: No need to provide any seperators
- it follows heirarchy like ['\n\n' -> '\n' -> , -> .]


```python
var=CharacterTextSplitter(
    separator='\n',
    chunk_size=100,
    chunk_overlap=20,
    is_separator_regex=False,
    length_function=len
)
```

```python
result= var.create_documents([Text])
result
```

```python
len(result)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    # No Need to provide any Separators
    # it follows heirarchy like ['\n\n' -> '\n' -> , -> .]
    chunk_size=100,
    chunk_overlap=0)
```

```python
responce=text_splitter.create_documents([Text])
responce # it automatically choosed '\n'
```

## How to label `METADATA`

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader(r"/content/Chidvilas_Kumkapalla_Resume.pdf")
restext=loader.load()
restext
```

```python
len(restext)
```

### So originally 2 docs was present , for total coverage we must add two docs mannually for creating docs using text splitter

```python
# TEXT SPLITTING
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=0)
```

```python
text1=restext[0].page_content
text2=restext[1].page_content
```

```python
response=text_splitter.create_documents([text1,text2])
response
```

```python
len(response)
```

```python
# So NO metadata
# now add metadata manually

metadata=[{'page': 1}, {'page': 2}]
```

```python
Newdocs=text_splitter.create_documents([text1,text2],
                                       metadatas=metadata)
```
