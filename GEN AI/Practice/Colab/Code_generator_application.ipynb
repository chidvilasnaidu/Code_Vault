```python
from langchain_core.prompts import HumanMessagePromptTemplate, SystemMessagePromptTemplate, ChatPromptTemplate,
```

```python
user_temp="Hello Bot, Can you please able to create {topic} Web application, Using HTML,CSS,JS"
system_msg= ''' Hello, system Behave like a experienced Developer ,You need to produce the codes based on human prompt, you must crosscheck before producing the code
if  the User ask About python give like
--- app.py---
"
"
--- ---

if he ask about Web Page or Application

--- Index.html---
"

"
--- ---
Likewise CSS code
--- Index.CSS---
"

"
LikeWise JAVASCRIPT code
--- ---
--- Index.JS---
"

"
--- ---
'''
```

```python
user_msg= HumanMessagePromptTemplate.from_template(user_temp)
Prompt_var= ChatPromptTemplate.from_messages([system_msg, user_msg])
```

```python
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
import os
key= userdata.get('Gemini')
os.environ['GOOGLE_API_KEY']=key
```

```python
Model=ChatGoogleGenerativeAI(
    model="gemini-2.5-flash")
Model
```

```python
from langchain_core.output_parsers import StrOutputParser
parser=StrOutputParser()
parser
```

```python
# LECL Chain

chain=Prompt_var|Model|parser
response=chain.invoke({
    "topic":"Relastic Clock with stop watch, timer, timeZones  "
})
```
