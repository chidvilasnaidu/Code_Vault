## Basic Multi-turn conversation System

- The below Model is basic and works only with in the colab
- If you want a system that will provides interface then we need to allocate some memory to the system , Either in local storage or compyter

```python
from google.colab import userdata
from google import genai
import os
key=userdata.get('Mutiturnmemory_key')
os.environ['GEMINI_API_KEY']=key
client=genai.Client()
```

```python
systemPromt=" Assume like you are a master or expert in Data Science and ML ,So give the answers in One lines"
chat = client.chats.create(

    model="gemini-3.5-flash",
    config=genai.types.GenerateContentConfig(
        system_instruction=systemPromt
    )
)
print("Hello Iam Medha!..., You can ask anything .... :)")
while True:
  Query=input(" Enter query...... ")
  if Query.casefold() in ['bye', 'over','quit','thank you']:
    print(" Fell free to ask !!, Thank you ... Bye Bye..")
    break
  else:
    response = chat.send_message(Query)
    print(response.text)
```

### In above code We maintained the meomory by executing the loop
