# Prompt Templates in LangChain

## 1. Prompt Template
- Prompt Template is used to create a single text prompt with dynamic variables.
- It generates a plain string as output.
- It is suitable for simple tasks and non-chat models.

## 2. Chat Prompt Template
- Chat Prompt Template is designed for chat-based models like GPT, Claude, and Gemini.
- It organizes prompts into messages with roles such as system, human, and AI.
- It helps manage context and multi-turn conversations effectively.

## 3. Few-Shot Messages Prompt Template
- Few-Shot Messages Prompt Template provides example interactions before the actual query.
- It improves response accuracy and consistency by demonstrating the desired pattern.
- It is useful for tasks such as classification, formatting, and reasoning.

### `partial_variables` Parameter
- it was used to set Default arguments or parameters
- they will assigned by computer to model whenever user input is misssing

```python
template=" Hi, Can you please Explain About {Text1} topic indetail in {num} Sentences"
```

### Chat Prompt Templates

```python
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate, AIMessagePromptTemplate, ChatPromptTemplate
```

```python
Systemtemp=" When Ever I ask Query give it point wise in {num} lines"
humantemp=" Explain about {topic}"
```

###Few-Shot Messages Prompt Template

```python
# Few-Shot Example by using Prompt templates methods
from google.colab import userdata
import os
from langchain_google_genai import ChatGoogleGenerativeAI
Gen_Key= userdata.get("GEMINI3")
os.environ['GOOGLE_API_KEY']=Gen_Key
Model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash")
```

```python
systemp=" you are expec in solving logical and reasoning problems, so give indetail Explaination"
humantemp=" what is ans of {op1}+{op2}"
# AItemp=" Answer is {ans}"
```

```python
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate, AIMessagePromptTemplate, ChatPromptTemplate
```

```python
sysmessage=SystemMessagePromptTemplate.from_template(systemp)
humanmsg=HumanMessagePromptTemplate.from_template(humantemp)
#AItemp=AIMessagePromptTemplate.from_template(AItemp)
```

```python
messages=ChatPromptTemplate.from_messages([sysmessage,humanmsg])
```

```python
prompt=messages.invoke({
    "op1" : 'A',
    "op2" : 'B',
})
```

```python
response=Model.invoke(prompt)
print(response.content)
```

 Few-Shot Example by using Prompt Chain  methods

```python
msg={
    "op1" : 'A',
    "op2" : 'B',
}
```

```python
chain= messages|Model
res=chain.invoke(msg)
print(res.content)
```

### Using direct funtion `few_shot_with_templates`

```python

from google.colab import userdata
import os
from langchain_google_genai import ChatGoogleGenerativeAI
Gen_Key= userdata.get("GEMINI4")
os.environ['GOOGLE_API_KEY']=Gen_Key
Model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash")
```

```python
systemp=" you are expec in solving logical and reasoning problems, so give indetail Explaination"
humantemp=" what is ans of {op1} + {op2}"
AItemp=" Answer is {ans}"
```

```python
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate, AIMessagePromptTemplate, ChatPromptTemplate
sysmessage=SystemMessagePromptTemplate.from_template(systemp)
humanmsg=HumanMessagePromptTemplate.from_template(humantemp)
AItemp=AIMessagePromptTemplate.from_template(AItemp)
```

```python
messages=ChatPromptTemplate.from_messages([sysmessage,humanmsg,AItemp])
```

```python
ex=[
    { "op1": 'A',"op2": 'B',"ans": 30 },
    { "op1": 'B',"op2": 'C',"ans": 50 },
    { "op1": 'C',"op2": 'D',"ans": 70 }
]
```

```python
from langchain_core.prompts import FewShotChatMessagePromptTemplate
```

```python
fewshot=FewShotChatMessagePromptTemplate(
    examples=ex,
    example_prompt=messages,
    input_variables=['op1','op2']

)
fewshot
```

```python
prompt_val=ChatPromptTemplate.from_messages([fewshot, humanmsg])
```

```python
#Human msg already created
chain= prompt_val|Model|parser
res=chain.invoke({
    'op1': "D",
    'op2': "E"
})
print(res)
```
