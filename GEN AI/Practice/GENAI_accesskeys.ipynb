```python

from groq import Groq
```

```python
client = Groq()
```

```python
UserInput=input("Enter Your Question : ")
systemprompt="hey Bot Make sure to give in simple and understandale format , Not extending 150 words , use lists if only required "


chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": UserInput
        },
          {
            "role": "system",
            "content": systemprompt # systemprompt used so say instructions to AI mo
        }
    ],
    model="llama-3.3-70b-versatile",
)
```

```python
chat_completion.choices[0].message.content
```

```python
UserInput=input("Enter Your Question : ")
systemprompt="hey Bot Make sure to give in simple and understandale format , Not extending 150 words , use lists if only required "

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": UserInput
        },
          {
            "role": "system",
            "content": systemprompt
        }
    ],
    model="llama-3.3-70b-versatile",
)
```

```python
chat_completion
```

```python
while True:
  UserInput=input("Enter Your Question : ")
  systemprompt="hey Bot Make sure to give in simple and understandale format , Not extending 150 to 250words , use lists if only required "

  if UserInput.lower() in ['over','bye', 'thank you','okay','quit','quit()']:
    print("Thank you!, Have a Great Day")
    break
  else:
    chat_completion = client.chat.completions.create(
      messages=[
          {
              "role": "user",
              "content": UserInput
          },
            {
              "role": "system",
              "content": systemprompt
          }
      ],
      model="llama-3.3-70b-versatile",
      )
    print(chat_completion.choices[0].message.content)

```
