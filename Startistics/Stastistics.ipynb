```python
taxi=sb.load_dataset('taxis')
taxi
```

```python
plt.figure(figsize=(12,5))
sb.histplot(taxi.distance,kde=True)
plt.axvline(taxi.distance.mean(),color='Black',ls='-.',label='mean')
plt.axvline(taxi.distance.median(),color='Red',ls='-.',label='median')
plt.title(" Mean and median with outliers ")
plt.legend()
plt.show();
```

```python
#Identifing ouliers by IQR visual method >75 and <25 are considered as outliers 
taxi.distance.describe()
```

```python
taxi_without_Outliers=taxi.distance[(taxi.distance>0.980000)&(taxi.distance<3.210000)]
taxi_without_Outliers
```

```python
plt.figure(figsize=(15,5))
sb.histplot(taxi_without_Outliers,kde=True)
plt.axvline(taxi_without_Outliers.mean(),color='Red',ls='-.',label='mean')
plt.axvline(taxi_without_Outliers.median(),color='Black',ls='-.',label='median')
plt.xticks(range(0,10,2))
plt.title(" Mean and median with taxi_without_Outliers")
plt.legend()
plt.show();
```

```python
plt.figure(figsize=(15,5))
sb.histplot(taxi.distance,kde=True)
plt.axvline(taxi.distance.mean(),color='Red',ls='-.',label='mean')
plt.axvline(taxi.distance.median(),color='Black',ls='-.',label='median')
plt.xticks(range(0,10,2))
plt.title(" Mean and median with outliers ")
plt.legend()
plt.show();
```

```python
taxi.distance.mean()
taxi_without_Outliers.median()
```

## By Removing the outliers the Mean line comes towards the medain 
### like mean is decreased to  3.02 ->1.64

### moving outliers reduces skewness and moves the mean closer to the median, making it more representative of the dataset’s true center

- More Outliers :Makes mean streched towords right hand side
- no Outliers :Makes mean streched towords left
